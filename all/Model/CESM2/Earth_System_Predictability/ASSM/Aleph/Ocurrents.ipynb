```python
def pop_add_cyclic(da):
    #copied from Mattew Longs github (https://gist.github.com/matt-long/50433da346da8ac17cde926eec90a87c) and modified for DA 
    #This doesn't work with Dask (DA should be loaded)
    #Input DA must have the name and coords (TLONG and TLAT), and don't have other dims (time etc.)
    
    nj = pop_grid_io.TLAT.shape[0]
    ni = pop_grid_io.TLONG.shape[1]

    xL = int(ni/2 - 1)
    xR = int(xL + ni)

    tlon = pop_grid_io.TLONG.data
    tlat = pop_grid_io.TLAT.data
    
    tlon = np.where(np.greater_equal(tlon, min(tlon[:,0])), tlon-360., tlon)    
    lon  = np.concatenate((tlon, tlon + 360.), 1)
    lon = lon[:, xL:xR]

    if ni == 320:
        lon[367:-3, 0] = lon[367:-3, 0] + 360.        
    lon = lon - 360.
    
    lon = np.hstack((lon, lon[:, 0:1] + 360.))
    if ni == 320:
        lon[367:, -1] = lon[367:, -1] - 360.

    #-- trick cartopy into doing the right thing:
    #   it gets confused when the cyclic coords are identical
    lon[:, 0] = lon[:, 0] - 1e-8

    #-- periodicity
    lat = np.concatenate((tlat, tlat), 1)
    lat = lat[:, xL:xR]
    lat = np.hstack((lat, lat[:,0:1]))

    TLAT = xr.DataArray(lat, dims=('nlat', 'nlon'))
    TLONG = xr.DataArray(lon, dims=('nlat', 'nlon'))
    

    # copy vars
    v = da.name
    v_dims = da.dims
    
    # determine and sort other dimensions
    other_dims = set(v_dims) - {'nlat', 'nlon'}
    other_dims = tuple([d for d in v_dims if d in other_dims])
    lon_dim = da.dims.index('nlon')
    field = da.data
    field = np.concatenate((field, field), lon_dim)
    field = field[..., :, xL:xR]
    field = np.concatenate((field, field[..., :, 0:1]), lon_dim)       
    dao = xr.DataArray(field, dims=other_dims+('nlat', 'nlon'), attrs=da.attrs)
    dao = dao.assign_coords(TLONG=TLONG)
    dao = dao.assign_coords(TLAT=TLAT)

            
    return dao
```

```python
exceptcv=['time','nlat','nlon','z_t',*VAR]
def process_coords(ds, drop=True, except_coord_vars=exceptcv):
    """Preprocessor function to drop all non-dim coords, which slows down concatenation."""
    coord_vars = []
    for v in np.array(ds.coords) :
        if not v in except_coord_vars:
            coord_vars += [v]
    for v in np.array(ds.data_vars) :
        if not v in except_coord_vars:
            coord_vars += [v]
    ds=ds.isel(z_t=slice(0,10)).isel(nlon=slice(62,143), nlat=slice(82,275))
    return ds.drop(coord_vars)
```

```python
#ds_t=xr.open_dataset('/proj/jedwards/archive/b.e21.BHISTcmip6.f09_g17.LE2-1001.001/ocn/proc/tseries/month_1/b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.TEMP.185001-185912.nc')
ds_t=xr.open_dataset('/proj/ryamaguchi/00_prgs/xgcm/b.e21.BHISTcmip6.f09_g17.LE2-1001.001.pop.h.TEMP.185001-185912.nc')
#ds_t = ds_t.TEMP.isel(nlon=slice(70,142), nlat=slice(82,273), z_t=slice(0,11))
#nlat = ds.lat.isel(nlat=slice(82,273))
#nlon = ds.lon.isel(nlat=slice(70,142))

ds["DZT"] = (xr.full_like(pop_grid_io.DXT,fill_value=1.)*pop_grid_io.dz).transpose('z_t','nlat','nlon')
ds["DZU"] = (xr.full_like(pop_grid_io.DXU,fill_value=1.)*pop_grid_io.dz).transpose('z_t','nlat','nlon')
ds["DXT"] = pop_grid_io.DXT
ds["DYT"] = pop_grid_io.DYT
ds["DXU"] = pop_grid_io.DXU
ds["DYU"] = pop_grid_io.DYU
ds["UAREA"] = pop_grid_io.UAREA
ds["TAREA"] = pop_grid_io.TAREA
ds["dz"] = pop_grid_io.dz
ds["hflux_factor"]=ds_t.hflux_factor

ds.DZT.attrs["long_name"] = "Thickness of T cells"
ds.DZT.attrs["units"] = "centimeter"
ds.DZT.attrs["grid_loc"] = "3111"
ds.DZU.attrs["long_name"] = "Thickness of U cells"
ds.DZU.attrs["units"] = "centimeter"
ds.DZU.attrs["grid_loc"] = "3221"

# make sure we have the cell volumne for calculations
VOL = (ds.DZT * ds.DXT * ds.DYT).compute()
KMT = pop_grid_io.KMT.compute()

for j in tqdm_notebook(range(len(KMT.nlat))):
    for i in range(len(KMT.nlon)):
        k = KMT.values[j, i].astype(int)
        VOL.values[k:, j, i] = 0.0

ds["VOL"] = VOL
ds.VOL.attrs["long_name"] = "volume of T cells"
ds.VOL.attrs["units"] = "centimeter^3"
ds.VOL.attrs["grid_loc"] = "3111"

```

```python
for i in tqdm_notebook(range(len(members))):
    
    budget = xr.Dataset()
    
    metrics = {("X",): ["DXU", "DXT"],("Y",): ["DYU", "DYT"],  
               ("Z",): ["DZU", "DZT"],("X", "Y"): ["UAREA", "TAREA"]}
    gridxgcm, dsxgcm = pop_tools.to_xgcm_grid_dataset(ds.isel(concat_dim=i),
                                                      periodic=True, metrics=metrics,
                                                      boundary={"X": "extend", "Y": "extend", "Z": "extend"})
    for coord in ["nlat", "nlon"]:
        if coord in dsxgcm.coords:
            dsxgcm = dsxgcm.drop_vars(coord)
    ## Tendency
    budget['UVEL'] = gridxgcm.diff(dsxgcm.UVEL, axis="X")
    budget['VVEL'] = gridxgcm.diff(dsxgcm.VVEL, axis="Y")
    
    if i==0: budgets=budget
    if i!=0: budgets=xr.concat([budgets,budget],dim='concat_dim')
```
