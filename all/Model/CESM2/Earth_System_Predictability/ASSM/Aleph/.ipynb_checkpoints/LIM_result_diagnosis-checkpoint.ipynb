```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 0. Settings
# ==========================================================
ens = 0
max_lag = 24
lags = np.arange(0, max_lag + 1)

state_names = ["PP", "SST", "SSH", "MLD"]

# Hindcast experiments to compare
Y_names = [
    "Y_CS_CW",
    "Y_CS_White",
    "Y_CS_Colored",
    "Y_ST_CW",
    "Y_ST_White",
    "Y_ST_Colored",
]

colors  = ["tab:red", "tab:blue", "tab:green", "tab:orange", "tab:purple", "tab:brown"]
markers = ["o", "s", "^", "D", "v", "P"]

# ==========================================================
# 1. Skill metric
# ==========================================================
def acc(x, y, min_valid=10):
    valid = np.isfinite(x) & np.isfinite(y)
    if valid.sum() < min_valid:
        return np.nan
    return np.corrcoef(x[valid], y[valid])[0, 1]

# ==========================================================
# 2. Prepare truth (xvec)
# ==========================================================
xvec = ds["xvec"].isel(ens_LE=ens).values  # (time, state)
nt_full, n_state = xvec.shape

# ==========================================================
# 3. Persistence skill (baseline)
# ==========================================================
skill_persist = np.full((n_state, max_lag + 1), np.nan)

for s in range(n_state):
    for tau in range(1, max_lag + 1):
        skill_persist[s, tau] = acc(
            xvec[:-tau, s],   # persistence forecast
            xvec[tau:,  s]    # truth
        )

# ==========================================================
# 4. Hindcast skills
# ==========================================================
skill_hind = {
    name: np.full((n_state, max_lag + 1), np.nan)
    for name in Y_names
}

for name in Y_names:

    Y = (
        ds[name]
        .isel(ens_LE=ens)
        .transpose("init_year", "month", "lead", "state")
    )

    n_init  = Y.sizes["init_year"]
    n_month = Y.sizes["month"]
    n_lead  = Y.sizes["lead"]

    # (time, lead, state)
    Y_flat = Y.values.reshape(n_init * n_month, n_lead, n_state)

    nt = min(nt_full, Y_flat.shape[0])

    for s in range(n_state):
        for tau in range(1, max_lag + 1):
            lead_idx = tau - 1
            if lead_idx < n_lead and tau < nt:
                skill_hind[name][s, tau] = acc(
                    xvec[tau:nt, s],                     # truth: x(t+tau)
                    Y_flat[:nt - tau, lead_idx, s]       # pred : Y(t, tau)
                )

# ==========================================================
# 5. Plot: 4 panels (state-wise)
# ==========================================================
fig, axes = plt.subplots(
    2, 2,
    figsize=(11, 8),
    sharex=True,
    sharey=True
)
axes = axes.flatten()

for s, ax in enumerate(axes):

    # --------------------------
    # Persistence (baseline)
    # --------------------------
    ax.plot(
        lags,
        skill_persist[s],
        color="gray",
        lw=2.0,
        linestyle="--",
        label="Persistence"
    )

    # --------------------------
    # Hindcast experiments
    # --------------------------
    for i, name in enumerate(Y_names):
        ax.plot(
            lags,
            skill_hind[name][s],
            color=colors[i],
            marker=markers[i],
            lw=1.8,
            ms=5,
            label=name.replace("Y_", "")
        )

    ax.axhline(0, color="k", lw=0.5)
    ax.set_xlim(0, max_lag)
    ax.set_title(state_names[s])

    if s % 2 == 0:
        ax.set_ylabel("ACC")
    if s >= 2:
        ax.set_xlabel("Lead (months)")

# Legend (single, clean)
axes[1].legend(
    frameon=False,
    fontsize=9,
    ncol=2
)

fig.suptitle(
    "Prediction skill by state\nPersistence vs LIM hindcasts",
    fontsize=14
)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 0. Settings
# ==========================================================
max_lag = 24
lags = np.arange(0, max_lag + 1)

state_names = ["PP", "SST", "SSH", "MLD"]

Y_names = [
    "Y_CS_CW",
    "Y_CS_White",
    "Y_CS_Colored",
    "Y_ST_CW",
    "Y_ST_White",
    "Y_ST_Colored",
]

colors  = ["tab:red", "tab:blue", "tab:green", "tab:orange", "tab:purple", "tab:brown"]
markers = ["o", "s", "^", "D", "v", "P"]

# ==========================================================
# 1. Skill metric
# ==========================================================
def acc(x, y, min_valid=10):
    valid = np.isfinite(x) & np.isfinite(y)
    if valid.sum() < min_valid:
        return np.nan
    return np.corrcoef(x[valid], y[valid])[0, 1]

# ==========================================================
# 2. Truth: ensemble-wise
# ==========================================================
xvec_all = ds["xvec"].values          # (ens, time, state)
n_ens, nt_full, n_state = xvec_all.shape

# ==========================================================
# 3. Persistence skill (ensemble-wise)
# ==========================================================
skill_persist = np.full((n_ens, n_state, max_lag + 1), np.nan)

for e in range(n_ens):
    for s in range(n_state):
        for tau in range(1, max_lag + 1):
            skill_persist[e, s, tau] = acc(
                xvec_all[e, :-tau, s],
                xvec_all[e,  tau:, s]
            )

# ==========================================================
# 4. Hindcast skill (ensemble-wise, corresponding truth)
# ==========================================================
skill_hind = {
    name: np.full((n_ens, n_state, max_lag + 1), np.nan)
    for name in Y_names
}

for e in range(n_ens):

    x_truth = xvec_all[e]   # (time, state)

    for name in Y_names:

        Y = (
            ds[name]
            .isel(ens_LE=e)
            .transpose("init_year", "month", "lead", "state")
        )

        n_init  = Y.sizes["init_year"]
        n_month = Y.sizes["month"]
        n_lead  = Y.sizes["lead"]

        # flatten (time, lead, state)
        Y_flat = Y.values.reshape(n_init * n_month, n_lead, n_state)

        nt = min(nt_full, Y_flat.shape[0])

        for s in range(n_state):
            for tau in range(1, max_lag + 1):
                lead_idx = tau - 1
                if lead_idx < n_lead and tau < nt:
                    skill_hind[name][e, s, tau] = acc(
                        x_truth[tau:nt, s],              # truth(e)
                        Y_flat[:nt - tau, lead_idx, s]  # hindcast(e)
                    )

# ==========================================================
# 5. Plot: ensemble mean ± spread
# ==========================================================
fig, axes = plt.subplots(
    2, 2,
    figsize=(11, 8),
    sharex=True,
    sharey=True
)
axes = axes.flatten()

for s, ax in enumerate(axes):

    # --------------------------
    # Persistence (mean ± std)
    # --------------------------
    mean_p = np.nanmean(skill_persist[:, s, :], axis=0)
    std_p  = np.nanstd (skill_persist[:, s, :], axis=0)

    ax.fill_between(
        lags,
        mean_p - std_p,
        mean_p + std_p,
        color="gray",
        alpha=0.15,
        linewidth=0
    )
    ax.plot(
        lags,
        mean_p,
        color="gray",
        lw=2.0,
        linestyle="--",
        label="Persistence"
    )

    # --------------------------
    # Hindcast experiments
    # --------------------------
    for i, name in enumerate(Y_names):

        mean_h = np.nanmean(skill_hind[name][:, s, :], axis=0)
        std_h  = np.nanstd (skill_hind[name][:, s, :], axis=0)

        ax.fill_between(
            lags,
            mean_h - std_h,
            mean_h + std_h,
            color=colors[i],
            alpha=0.15,
            linewidth=0
        )

        ax.plot(
            lags,
            mean_h,
            color=colors[i],
            marker=markers[i],
            lw=1.8,
            ms=5,
            label=name.replace("Y_", "")
        )

    ax.axhline(0, color="k", lw=0.5)
    ax.set_xlim(0, max_lag)
    ax.set_title(state_names[s])

    if s % 2 == 0:
        ax.set_ylabel("ACC")
    if s >= 2:
        ax.set_xlabel("Lead (months)")

# Legend in SST panel
axes[1].legend(
    frameon=False,
    fontsize=9,
    ncol=2
)

fig.suptitle(
    "Prediction skill by state\nPersistence vs LIM hindcasts (ensemble-consistent)",
    fontsize=14
)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 0. Settings
# ==========================================================
ens = 0
max_lag = 24
lags = np.arange(0, max_lag + 1)

# State definition
# state=0 : PP
# state=1 : SST
# state=2 : SSH
# state=3 : MLD
predictors = {
    1: "SST",
    2: "SSH",
    3: "MLD",
}

# ==========================================================
# 1. Training lead–lag correlation (xvec)
# ==========================================================
def lead_lag_corr_pos(x, y, max_lag):
    """ lag >= 0 only """
    r = np.full(max_lag + 1, np.nan)

    for lag in range(max_lag + 1):
        if lag == 0:
            xx, yy = x, y
        else:
            xx, yy = x[:-lag], y[lag:]

        valid = np.isfinite(xx) & np.isfinite(yy)
        if valid.sum() >= 10:
            r[lag] = np.corrcoef(xx[valid], yy[valid])[0, 1]

    return r

# ==========================================================
# 2. Prepare data
# ==========================================================
xvec = ds["xvec"].isel(ens_LE=ens).values  # (time, state)

Y = (
    ds["Y_CS_CW"]
    .isel(ens_LE=ens)
    .transpose("init_year", "month", "lead", "state")
)

n_init  = Y.sizes["init_year"]
n_month = Y.sizes["month"]
n_lead  = Y.sizes["lead"]

Y_flat = Y.values.reshape(n_init * n_month, n_lead, Y.sizes["state"])
# shape: (time, lead, state)

# ==========================================================
# 3. Plot: 1 × 3 (overlay)
# ==========================================================
fig, axes = plt.subplots(
    1, 3,
    figsize=(14, 4),
    sharey=True
)

for j, (pidx, pname) in enumerate(predictors.items()):
    ax = axes[j]

    # ---------------------------
    # Training (xvec)
    # ---------------------------
    r_train = lead_lag_corr_pos(
        xvec[:, pidx],
        xvec[:, 0],
        max_lag
    )

    ax.plot(
        lags,
        r_train,
        color="k",
        lw=2.2,
        marker="o",
        markersize=5,
        label="Training (xvec)"
    )

    # ---------------------------
    # Hindcast (LIM)
    # ---------------------------
    r_hind = np.full(max_lag + 1, np.nan)

    for lag in range(1, max_lag + 1):
        lead_idx = lag - 1
        if lead_idx >= n_lead:
            continue
    
        nt = min(len(xvec), Y_flat.shape[0])
    
        x_ref = xvec[:nt, pidx]
        y_fc  = Y_flat[:nt, lead_idx, 0]
    
        valid = np.isfinite(x_ref) & np.isfinite(y_fc)
        if valid.sum() >= 10:
            r_hind[lag] = np.corrcoef(
                x_ref[valid],
                y_fc[valid]
            )[0, 1]


    # lag = 0 → training consistency
    r_hind[0] = r_train[0]

    ax.plot(
        lags,
        r_hind,
        color="tab:red",
        lw=2.0,
        marker="s",
        markersize=5,
        label="LIM hindcast"
    )

    # ---------------------------
    # Cosmetics
    # ---------------------------
    ax.axhline(0, color="gray", lw=0.6)
    ax.set_title(pname)
    ax.set_xlabel("Lag (months)")
    ax.set_xlim(0, max_lag)

    if j == 0:
        ax.set_ylabel("Correlation")
        ax.legend(frameon=False, fontsize=10)

fig.suptitle(
    "Lead–lag correlation (lag ≥ 0): predictors → PP",
    fontsize=14
)

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 0. Settings
# ==========================================================
ens = 0
max_lag = 24
lags = np.arange(0, max_lag + 1)

predictors = {
    1: "SST",
    2: "SSH",
    3: "MLD",
}

leadtau_list = [0, 6, 12, 18]

# ==========================================================
# 1. Lead–lag correlation
# ==========================================================
def lead_lag_corr_pos(x, y, max_lag):
    r = np.full(max_lag + 1, np.nan)

    for lag in range(max_lag + 1):
        if lag == 0:
            xx, yy = x, y
        else:
            xx, yy = x[:-lag], y[lag:]

        valid = np.isfinite(xx) & np.isfinite(yy)
        if valid.sum() >= 10:
            r[lag] = np.corrcoef(xx[valid], yy[valid])[0, 1]

    return r

# ==========================================================
# 2. Prepare data
# ==========================================================
xvec = ds["xvec"].isel(ens_LE=ens).values  # (time, state)

Y = (
    ds["Y_CS_CW"]
    .isel(ens_LE=ens)
    .transpose("init_year", "month", "lead", "state")
)

n_init  = Y.sizes["init_year"]
n_month = Y.sizes["month"]
n_lead  = Y.sizes["lead"]

Y_flat = Y.values.reshape(n_init * n_month, n_lead, Y.sizes["state"])

nt = min(len(xvec), Y_flat.shape[0])
t  = np.arange(nt)

# ==========================================================
# 3. Plot: 5 × 3
# ==========================================================
fig, axes = plt.subplots(
    1 + len(leadtau_list), 3,
    figsize=(14, 12)
)

for j, (pidx, pname) in enumerate(predictors.items()):

    # ======================================================
    # (1) Lead–lag correlation (top row)
    # ======================================================
    ax = axes[0, j]

    r_train = lead_lag_corr_pos(
        xvec[:, pidx],
        xvec[:, 0],
        max_lag
    )

    ax.plot(lags, r_train, "ko-", lw=2.2, ms=5, label="Training (xvec)")

    r_hind = np.full(max_lag + 1, np.nan)

    for lag in range(1, max_lag + 1):
        lead_idx = lag - 1
        if lead_idx >= n_lead:
            continue

        x_ref = xvec[:nt, pidx]
        y_fc  = Y_flat[:nt, lead_idx, 0]

        valid = np.isfinite(x_ref) & np.isfinite(y_fc)
        if valid.sum() >= 10:
            r_hind[lag] = np.corrcoef(x_ref[valid], y_fc[valid])[0, 1]

    r_hind[0] = r_train[0]

    ax.plot(lags, r_hind, "s-", color="tab:red", lw=2.0, ms=5, label="LIM hindcast")

    ax.axhline(0, color="gray", lw=0.6)
    ax.set_title(pname)
    # ax.set_xlim(0, max_lag)
    ax.set_xlim(0, max_lag)

    if j == 0:
        ax.set_ylabel("Correlation")
        ax.legend(frameon=False, fontsize=9)

    # ======================================================
    # (2–5) Time series rows
    # ======================================================
    for i, tau in enumerate(leadtau_list):
        ax = axes[i + 1, j]

        if tau == 0:
            y_pp = xvec[:nt, 0]   # PP at tau=0
        else:
            lead_idx = tau - 1
            if lead_idx >= n_lead:
                continue
            y_pp = Y_flat[:nt, lead_idx, 0]

        x_pred = xvec[:nt, pidx]

        ax.plot(t, x_pred, color="k", lw=1.0, label=f"{pname} (τ=0)")
        ax.plot(t, y_pp, color="tab:red", lw=1.0, label=f"PP (τ={tau})")

        ax.axhline(0, color="gray", lw=0.5)

        if j == 0:
            ax.set_ylabel(f"τ = {tau}")

        if i == len(leadtau_list) - 1:
            ax.set_xlabel("Time index")

        if j == 2 and i == 0:
            ax.legend(frameon=False, fontsize=9)
        ax.set_xlim(0, nt - 1)

# ==========================================================
# Final layout
# ==========================================================
fig.suptitle(
    "Lead–lag correlation and time series comparison\n"
    "Predictor (τ=0) vs PP (τ = 0, 6, 12, 18)",
    fontsize=14
)

plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from matplotlib.colors import ListedColormap

# ==========================================================
# 0. Settings
# ==========================================================
ens = 0
max_lag = 24
n_month = 12

predictors = {
    1: "SST",
    2: "SSH",
    3: "MLD",
}

month_names = ["Jan","Feb","Mar","Apr","May","Jun",
               "Jul","Aug","Sep","Oct","Nov","Dec"]

# ===== DOF (same as MATLAB) =====
N = 64
nu = N - 2
tcrit = t.ppf(1 - 0.05/2, nu)
rcrit = np.sqrt(tcrit**2 / (tcrit**2 + nu))

# ==========================================================
# 1. Prepare data
# ==========================================================
xvec = ds["xvec"].isel(ens_LE=ens).values   # (time, state)
nt, n_state = xvec.shape

# month index (assumes continuous monthly data)
month_idx = np.tile(np.arange(n_month), nt // n_month)

# ==========================================================
# 2. Compute lead–lag correlations (manual corr)
# ==========================================================
corr_maps = {}

# ---------- Cross correlations ----------
for pidx, pname in predictors.items():

    C = np.full((n_month, max_lag + 1), np.nan)

    for m in range(n_month):
        for lag in range(max_lag + 1):

            if lag == 0:
                tmask = month_idx == m
                x = xvec[tmask, pidx]
                y = xvec[tmask, 0]
            else:
                t0 = np.where(month_idx[:-lag] == m)[0]
                t1 = t0 + lag
                x = xvec[t0, pidx]
                y = xvec[t1, 0]

            valid = np.isfinite(x) & np.isfinite(y)
            if valid.sum() < 10:
                continue

            xv = x[valid]
            yv = y[valid]

            sx = np.nanstd(xv, ddof=1)
            sy = np.nanstd(yv, ddof=1)

            if sx > 0 and sy > 0:
                cov = np.nanmean((xv - xv.mean()) * (yv - yv.mean()))
                C[m, lag] = cov / (sx * sy)

    corr_maps[pname] = C

# ---------- Auto correlation: PP → PP ----------
C_auto = np.full((n_month, max_lag + 1), np.nan)

for m in range(n_month):
    for lag in range(max_lag + 1):

        if lag == 0:
            tmask = month_idx == m
            x = xvec[tmask, 0]
            y = xvec[tmask, 0]
        else:
            t0 = np.where(month_idx[:-lag] == m)[0]
            t1 = t0 + lag
            x = xvec[t0, 0]
            y = xvec[t1, 0]

        valid = np.isfinite(x) & np.isfinite(y)
        if valid.sum() < 10:
            continue

        xv = x[valid]
        yv = y[valid]

        sx = np.nanstd(xv, ddof=1)
        sy = np.nanstd(yv, ddof=1)

        if sx > 0 and sy > 0:
            cov = np.nanmean((xv - xv.mean()) * (yv - yv.mean()))
            C_auto[m, lag] = cov / (sx * sy)

corr_maps["PP"] = C_auto

# ==========================================================
# 3. MATLAB-style colormap
# ==========================================================
nColors = 20
base_colors = np.array([[0,0,1],[1,1,1],[1,0,0]])
x_base = np.linspace(-1, 1, 3)
x_new  = np.linspace(-1, 1, nColors)

cmap_arr = np.zeros((nColors, 3))
for i in range(3):
    cmap_arr[:, i] = np.interp(x_new, x_base, base_colors[:, i])

cmap = ListedColormap(cmap_arr)

# ==========================================================
# 4. Plot (2×2)
# ==========================================================
fig, axes = plt.subplots(2, 2, figsize=(12, 9), sharex=True, sharey=True)
axes = axes.flatten()

panel_order = ["SST", "SSH", "MLD", "PP"]

x_edges = np.arange(max_lag + 2)
y_edges = np.arange(n_month + 1)

for i, key in enumerate(panel_order):
    ax = axes[i]
    C = corr_maps[key]

    im = ax.pcolormesh(
        x_edges, y_edges, C,
        cmap=cmap, vmin=-1, vmax=1, shading="flat"
    )

    ax.set_xlim(0, max_lag + 1)
    ax.set_ylim(0, n_month)

    ax.set_yticks(np.arange(n_month) + 0.5)
    ax.set_yticklabels(month_names)
    ax.set_xticks(np.arange(max_lag + 1) + 0.5)
    ax.set_xticklabels(np.arange(max_lag + 1))

    ax.axvline(0, color="k", lw=0.8)

    if key == "PP":
        ax.set_title("NPP (t+τ) ← NPP (t)")
    else:
        ax.set_title(f"NPP (t+τ) ← {key} (t)")

    ax.set_ylabel("Initialization month")

    # significance dots
    sig = np.abs(C) >= rcrit
    iy, ix = np.where(sig)
    if len(ix) > 0:
        ax.scatter(ix + 0.5, iy + 0.5, s=40, c="k", marker=".")

axes[-1].set_xlabel("Lead (months)")

# ==========================================================
# Colorbar
# ==========================================================
cax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
cb = fig.colorbar(im, cax=cax)
cb.set_label("Correlation")
cb.set_ticks(np.linspace(-1, 1, 11))

fig.suptitle(
    "Lead–lag cross correlation from xvec only\n"
    r"$\mathrm{corr}(x(t), \mathrm{PP}(t+\tau))$",
    fontsize=14
)

plt.tight_layout(rect=[0, 0, 0.9, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from matplotlib.colors import ListedColormap

# ==========================================================
# 0. Settings
# ==========================================================
ens = 0
max_lag = 24
n_month = 12

predictors = {
    1: "SST",
    2: "SSH",
    3: "MLD",
}

month_names = ["Jan","Feb","Mar","Apr","May","Jun",
               "Jul","Aug","Sep","Oct","Nov","Dec"]

# ===== MATLAB과 동일한 DOF 설정 =====
N = 64
nu = N - 2
tcrit = t.ppf(1 - 0.05/2, nu)
rcrit = np.sqrt(tcrit**2 / (tcrit**2 + nu))

# ==========================================================
# 1. Prepare data
# ==========================================================
xvec = ds["xvec"].isel(ens_LE=ens).values   # (time, state)

Y = (
    ds["Y_CS_CW"]
    .isel(ens_LE=ens)
    .transpose("init_year", "month", "lead", "state")
)

n_init = Y.sizes["init_year"]
n_lead = Y.sizes["lead"]

Y_flat = Y.values.reshape(n_init * n_month, n_lead, Y.sizes["state"])
month_idx = np.tile(np.arange(n_month), n_init)

# ==========================================================
# 2. Month-wise PP(t) std (normalization reference)
# ==========================================================
pp0_std = np.full(n_month, np.nan)
for m in range(n_month):
    mask = month_idx == m
    pp0 = xvec[mask, 0]
    if np.isfinite(pp0).sum() >= 10:
        pp0_std[m] = np.nanstd(pp0, ddof=1)

# ==========================================================
# 3. Predictor → PP normalized covariance
# ==========================================================
corr_maps = {}

for pidx, pname in predictors.items():

    C = np.full((n_month, max_lag + 1), np.nan)

    for m in range(n_month):
        tmask = month_idx == m

        for lag in range(max_lag + 1):

            if lag == 0:
                x = xvec[tmask, pidx]
                y = xvec[tmask, 0]
            else:
                lead_idx = lag - 1
                if lead_idx >= n_lead:
                    continue
                x = xvec[tmask, pidx]
                y = Y_flat[tmask, lead_idx, 0]

            valid = np.isfinite(x) & np.isfinite(y)
            if valid.sum() < 10:
                continue

            xv = x[valid]
            yv = y[valid]

            std_x  = np.nanstd(xv, ddof=1)
            std_pp = pp0_std[m]

            if std_x > 0 and std_pp > 0:
                cov_xy = np.nanmean((xv - np.nanmean(xv)) *
                                     (yv - np.nanmean(yv)))
                C[m, lag] = cov_xy / (std_x * std_pp)

    corr_maps[pname] = C

# ==========================================================
# 4. Auto lead–lag (LIM): PP(t) → PP̂(t+τ)
# ==========================================================
C_auto = np.full((n_month, max_lag + 1), np.nan)

for m in range(n_month):
    tmask = month_idx == m
    pp0 = xvec[tmask, 0]

    for lag in range(max_lag + 1):

        if lag == 0:
            x = pp0
            y = pp0    # consistency at tau=0
        else:
            lead_idx = lag - 1
            if lead_idx >= n_lead:
                continue
            x = xvec[tmask, 0]                 # PP(t)
            y = Y_flat[tmask, lead_idx, 0]     # PP̂(t+τ) (hindcast)

        valid = np.isfinite(x) & np.isfinite(y)
        if valid.sum() < 10:
            continue

        xv = x[valid]
        yv = y[valid]

        std_pp = pp0_std[m]
        if std_pp > 0:
            cov_xy = np.nanmean(
                (xv - np.nanmean(xv)) *
                (yv - np.nanmean(yv))
            )
            C_auto[m, lag] = cov_xy / (std_pp * std_pp)


# ==========================================================
# 5. MATLAB-style colormap
# ==========================================================
nColors = 20
base_colors = np.array([[0,0,1],
                        [1,1,1],
                        [1,0,0]])
x_base = np.linspace(-1, 1, 3)
x_new  = np.linspace(-1, 1, nColors)

cmap_arr = np.zeros((nColors, 3))
for i in range(3):
    cmap_arr[:, i] = np.interp(x_new, x_base, base_colors[:, i])

cmap = ListedColormap(cmap_arr)

# ==========================================================
# 6. Plot (2×2)
# ==========================================================
fig, axes = plt.subplots(2, 2, figsize=(12, 9), sharex=True, sharey=True)
axes = axes.flatten()

vmin, vmax = -1, 1
x_edges = np.arange(max_lag + 2)
y_edges = np.arange(n_month + 1)

# --- first 3 panels: predictors ---
for i, (pname, C) in enumerate(corr_maps.items()):
    ax = axes[i]

    im = ax.pcolormesh(x_edges, y_edges, C,
                        cmap=cmap, vmin=vmin, vmax=vmax, shading="flat")

    ax.set_title(f"NPP (t+τ) ← {pname} (t)")
    ax.set_ylabel("Initialization month")

    sigmask = np.abs(C) >= rcrit
    iy, ix = np.where(sigmask)
    ax.scatter(ix + 0.5, iy + 0.5, s=40, c="k", marker=".", zorder=3)

# --- last panel: auto PP → PP ---
ax = axes[3]
im = ax.pcolormesh(x_edges, y_edges, C_auto,
                    cmap=cmap, vmin=vmin, vmax=vmax, shading="flat")

ax.set_title("NPP (t+τ) ← NPP (t)  (auto)")
sigmask = np.abs(C_auto) >= rcrit
iy, ix = np.where(sigmask)
ax.scatter(ix + 0.5, iy + 0.5, s=40, c="k", marker=".", zorder=3)

# --- common axis formatting ---
for ax in axes:
    ax.set_xlim(0, max_lag + 1)
    ax.set_ylim(0, n_month)
    ax.set_yticks(np.arange(n_month) + 0.5)
    ax.set_yticklabels(month_names)
    ax.set_xticks(np.arange(max_lag + 1) + 0.5)
    ax.set_xticklabels(np.arange(max_lag + 1))
    ax.axvline(0, color="k", lw=0.8)

axes[-1].set_xlabel("Lead (months)")

# ==========================================================
# Colorbar
# ==========================================================
cax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
cb = fig.colorbar(im, cax=cax)
cb.set_label("Normalized covariance")
cb.set_ticks(np.linspace(-1, 1, 11))

fig.suptitle(
    "Normalized lead–lag covariance (xvec only)\n"
    r"$\mathrm{cov}(\cdot,\cdot)/[\sigma_{\mathrm{PP}}(t)]$ reference",
    fontsize=14
)

plt.tight_layout(rect=[0, 0, 0.9, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# ==========================================================
# 0. Settings
# ==========================================================
ens = 0
max_lag = 24
n_month = 12

# state indices (xvec 기준)
states = {
    0: "NPP",
    1: "SST",
    2: "SSH",
    3: "MLD",
}

month_names = ["Jan","Feb","Mar","Apr","May","Jun",
               "Jul","Aug","Sep","Oct","Nov","Dec"]

# ==========================================================
# 1. Load data
# ==========================================================
# x(t)
xvec = ds["xvec"].isel(ens_LE=ens).values     # (time, state)

# LIM hindcast x(t+τ)
Y = (
    ds["Y_CS_CW"]
    .isel(ens_LE=ens)
    .transpose("init_year", "month", "lead", "state")
)

n_init = Y.sizes["init_year"]
n_lead = Y.sizes["lead"]

# reshape → (time, lead, state)
Y_flat = Y.values.reshape(n_init * n_month, n_lead, Y.sizes["state"])

# month index for θ
month_idx = np.tile(np.arange(n_month), n_init)

# ==========================================================
# 2. Correlation function K(θ,τ)
#     K_i(θ,τ) = < x_i(t+τ) x_i(t) >_θ
#     (NO mean removal, NO normalization)
# ==========================================================
K_maps = {}

for sidx, sname in states.items():

    K = np.full((n_month, max_lag + 1), np.nan)

    for m in range(n_month):
        tmask = month_idx == m

        for lag in range(max_lag + 1):

            if lag == 0:
                # τ = 0 → x(t) x(t)
                x0 = xvec[tmask, sidx]
                x1 = xvec[tmask, sidx]
            else:
                lead_idx = lag - 1
                if lead_idx >= n_lead:
                    continue

                # x(t), x(t+τ)
                x0 = xvec[tmask, sidx]
                x1 = Y_flat[tmask, lead_idx, sidx]

            valid = np.isfinite(x0) & np.isfinite(x1)
            if valid.sum() < 10:
                continue

            # 논문 정의 그대로: < x(t+τ) x(t) >
            K[m, lag] = np.nanmean(x1[valid] * x0[valid])

    K_maps[sname] = K

# ==========================================================
# 3. MATLAB-style colormap
# ==========================================================
nColors = 20
base_colors = np.array([[0,0,1],
                        [1,1,1],
                        [1,0,0]])
x_base = np.linspace(-1, 1, 3)
x_new  = np.linspace(-1, 1, nColors)

cmap_arr = np.zeros((nColors, 3))
for i in range(3):
    cmap_arr[:, i] = np.interp(x_new, x_base, base_colors[:, i])

cmap = ListedColormap(cmap_arr)

# ==========================================================
# 4. Plot (2×2) — 논문 K 스타일
# ==========================================================
fig, axes = plt.subplots(2, 2, figsize=(12, 9),
                         sharex=True, sharey=True)
axes = axes.flatten()

x_edges = np.arange(max_lag + 2)
y_edges = np.arange(n_month + 1)

vmax = np.nanmax([np.nanmax(np.abs(K)) for K in K_maps.values()])
vmin = -vmax

for i, (sname, K) in enumerate(K_maps.items()):
    ax = axes[i]

    im = ax.pcolormesh(
        x_edges, y_edges, K,
        cmap=cmap, vmin=vmin, vmax=vmax,
        shading="flat"
    )

    ax.set_title(rf"$K(\theta,\tau)$ : {sname}")
    ax.set_ylabel(r"$\theta$ (month)")

# axis formatting
for ax in axes:
    ax.set_xlim(0, max_lag + 1)
    ax.set_ylim(0, n_month)
    ax.set_yticks(np.arange(n_month) + 0.5)
    ax.set_yticklabels(month_names)
    ax.set_xticks(np.arange(max_lag + 1) + 0.5)
    ax.set_xticklabels(np.arange(max_lag + 1))
    ax.axvline(0, color="k", lw=0.8)

axes[-1].set_xlabel("Lead (months)")

# ==========================================================
# Colorbar
# ==========================================================
cax = fig.add_axes([0.92, 0.15, 0.02, 0.7])
cb = fig.colorbar(im, cax=cax)
cb.set_label(r"$\langle x(t+\tau)x(t)\rangle$")
cb.set_ticks(np.linspace(vmin, vmax, 9))

fig.suptitle(
    r"Correlation function $K(\theta,\tau)=\langle x(t+\tau)x(t)\rangle_\theta$",
    fontsize=14
)

plt.tight_layout(rect=[0, 0, 0.9, 0.95])
plt.show()
```
