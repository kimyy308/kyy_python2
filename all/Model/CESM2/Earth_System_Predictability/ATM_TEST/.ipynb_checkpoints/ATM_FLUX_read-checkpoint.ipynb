```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 1. Load DataArrays (★ Dataset → DataArray로 수정)
# ==========================================================
da1 = cfgs['TEMP'].A_F_REF1_ds
da2 = cfgs['TEMP'].A_F_REF2_ds

# 안전하게 time 정렬
da1, da2 = xr.align(da1, da2, join='inner')

# ==========================================================
# 2. Remove seasonal cycle (monthly climatology)
# ==========================================================
da1_anom = da1.groupby('time.month') - da1.groupby('time.month').mean('time')
da2_anom = da2.groupby('time.month') - da2.groupby('time.month').mean('time')

# ==========================================================
# 3. Remove linear trend (along time)
# ==========================================================
def detrend(da, dim='time'):
    p = da.polyfit(dim=dim, deg=1)
    trend = xr.polyval(da[dim], p.polyfit_coefficients)
    return da - trend

da1_dt = detrend(da1_anom)
da2_dt = detrend(da2_anom)

# ==========================================================
# 4. Correlations (surface only)
# ==========================================================
corr_raw = xr.corr(da1, da2, dim='time').isel(z_t=0)
corr_ds  = xr.corr(da1_anom, da2_anom, dim='time').isel(z_t=0)
corr_dt  = xr.corr(da1_dt, da2_dt, dim='time').isel(z_t=0)

# ==========================================================
# 5. Plot: 1 × 3 panel
# ==========================================================
fig, axes = plt.subplots(
    nrows=3, ncols=1,
    figsize=(6, 8),
    constrained_layout=True
)

titles = [
    'Raw',
    'Deseasoned',
    'Deseasoned + Detrended'
]

corrs = [corr_raw, corr_ds, corr_dt]

for ax, corr, title in zip(axes, corrs, titles):
    pcm = ax.pcolormesh(
        da1.lon, da1.lat, corr,
        cmap='RdBu_r',
        vmin=-1, vmax=1
    )
    ax.set_title(title)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

# shared colorbar
cbar = fig.colorbar(pcm, ax=axes, shrink=0.9)
cbar.set_label('Correlation')

plt.show()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 1. Load DataArrays
# ==========================================================
da1 = cfgs['TEMP'].A_F_REF1_ds
da2 = cfgs['TEMP'].A_F_REF2_ds

# time 정렬
da1, da2 = xr.align(da1, da2, join='inner')

# ==========================================================
# 2. Seasonal cycle from REF1 ONLY
# ==========================================================
seasonal_ref1 = da1.groupby('time.month').mean('time')

da1_ds = da1.groupby('time.month') - seasonal_ref1
da2_ds = da2.groupby('time.month') - seasonal_ref1   # ★ same signal removed

# ==========================================================
# 3. Linear trend from REF1 ONLY
# ==========================================================
def detrend_with_ref(ref, target, dim='time'):
    """
    Remove linear trend estimated from ref,
    applied to both ref and target
    """
    p = ref.polyfit(dim=dim, deg=1)
    trend = xr.polyval(ref[dim], p.polyfit_coefficients)
    return ref - trend, target - trend

da1_dt, da2_dt = detrend_with_ref(da1_ds, da2_ds)

# ==========================================================
# 4. Correlations (surface only)
# ==========================================================
corr_raw = xr.corr(da1, da2, dim='time').isel(z_t=0)
corr_ds  = xr.corr(da1_ds, da2_ds, dim='time').isel(z_t=0)
corr_dt  = xr.corr(da1_dt, da2_dt, dim='time').isel(z_t=0)

# ==========================================================
# 5. Plot: 3 × 1 panel
# ==========================================================
fig, axes = plt.subplots(
    nrows=3, ncols=1,
    figsize=(6, 8),
    constrained_layout=True
)

titles = [
    'Raw',
    'Deseasoned (REF1-based)',
    'Deseasoned + Detrended (REF1-based)'
]

corrs = [corr_raw, corr_ds, corr_dt]

for ax, corr, title in zip(axes, corrs, titles):
    pcm = ax.pcolormesh(
        corr,
        cmap='RdBu_r',
        vmin=-1, vmax=1
    )
    ax.set_title(title)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

# shared colorbar
cbar = fig.colorbar(pcm, ax=axes, shrink=0.9)
cbar.set_label('Correlation')

plt.show()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 1. Load DataArrays
# ==========================================================
da1 = cfgs['TEMP'].A_F_REF1_ds
da2 = cfgs['TEMP'].A_F_REF2_ds

# time 정렬
da1, da2 = xr.align(da1, da2, join='inner')

# ==========================================================
# 2. Remove seasonal cycle (monthly climatology)
# ==========================================================
da1_anom = da1.groupby('time.month') - da1.groupby('time.month').mean('time')
da2_anom = da2.groupby('time.month') - da2.groupby('time.month').mean('time')

# ==========================================================
# 3. Remove linear trend
# ==========================================================
def detrend(da, dim='time'):
    p = da.polyfit(dim=dim, deg=1)
    trend = xr.polyval(da[dim], p.polyfit_coefficients)
    return da - trend

da1_dt = detrend(da1_anom)
da2_dt = detrend(da2_anom)

# ==========================================================
# 4. Annual mean
# ==========================================================
da1_ann = da1.groupby('time.year').mean('time').rename({'year': 'time'})
da2_ann = da2.groupby('time.year').mean('time').rename({'year': 'time'})

# ==========================================================
# 5. RMSE (surface only)
# ==========================================================
rmse_raw = np.sqrt(((da1 - da2)**2).mean('time')).isel(z_t=0)
rmse_ds  = np.sqrt(((da1_anom - da2_anom)**2).mean('time')).isel(z_t=0)
rmse_dt  = np.sqrt(((da1_dt - da2_dt)**2).mean('time')).isel(z_t=0)
rmse_ann = np.sqrt(((da1_ann - da2_ann)**2).mean('time')).isel(z_t=0)

# ==========================================================
# 6. Plot: 2 × 2 panel
# ==========================================================
fig, axes = plt.subplots(
    nrows=2, ncols=2,
    figsize=(10, 6),
    constrained_layout=True
)

titles = [
    'Raw RMSE',
    'Deseasoned RMSE',
    'Deseasoned + Detrended RMSE',
    'Annual Mean RMSE'
]

rmses = [rmse_raw, rmse_ds, rmse_dt, rmse_ann]

for ax, rmse, title in zip(axes.flat, rmses, titles):
    pcm = ax.pcolormesh(
        da1.lon, da1.lat, rmse,
        cmap='viridis',   # RMSE는 sequential colormap 추천
        vmin=0, vmax=0.8
    )
    ax.set_title(title)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')

# shared colorbar
cbar = fig.colorbar(pcm, ax=axes, shrink=0.9, extend='max')
cbar.set_label('RMSE')

plt.show()
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# 1. Load DataArrays
# ==========================================================
da1 = cfgs['TEMP'].A_F_REF1_ds
da2 = cfgs['TEMP'].A_F_REF2_ds

# time 정렬
da1, da2 = xr.align(da1, da2, join='inner')

# ==========================================================
# 2. Deseason only (for maps & time series)
# ==========================================================
da1_anom = da1.groupby('time.month') - da1.groupby('time.month').mean('time')
da2_anom = da2.groupby('time.month') - da2.groupby('time.month').mean('time')

# ==========================================================
# 3. Raw RMSE & Raw Correlation (surface)
# ==========================================================
rmse_raw = np.sqrt(((da1 - da2) ** 2).mean('time')).isel(z_t=0)
corr_raw = xr.corr(da1, da2, dim='time').isel(z_t=0)

# ==========================================================
# 4. Points for time series (lon, lat)
# ==========================================================
points = [
    (-40, 40),
    (-40, 20),
    (-10, 70),
    (20, 60),
]

# ==========================================================
# 5. Extract deseasoned time series at points
# ==========================================================
ts1 = []
ts2 = []

for lon0, lat0 in points:
    da1_p = da1_anom.sel(lon=lon0, lat=lat0, method='nearest').isel(z_t=0)
    da2_p = da2_anom.sel(lon=lon0, lat=lat0, method='nearest').isel(z_t=0)
    ts1.append(da1_p)
    ts2.append(da2_p)

# ==========================================================
# 6. Plot: 3 × 2 panel
# ==========================================================
fig, axes = plt.subplots(
    nrows=3, ncols=2,
    figsize=(12, 10),
    constrained_layout=True
)

# --------------------------
# (1,1) Raw RMSE map
# --------------------------
ax = axes[0, 0]
pcm1 = ax.pcolormesh(
    da1.lon, da1.lat, rmse_raw,
    cmap='viridis',
    vmin=0, vmax=0.8
)
ax.set_title('Raw RMSE')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# mark points
for lon0, lat0 in points:
    ax.plot(lon0, lat0, marker='*', color='red', markersize=10)

# --------------------------
# (1,2) Raw Correlation map
# --------------------------
ax = axes[0, 1]
pcm2 = ax.pcolormesh(
    da1.lon, da1.lat, corr_raw,
    cmap='RdBu_r',
    vmin=-1, vmax=1
)
ax.set_title('Raw Correlation')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# mark points
for lon0, lat0 in points:
    ax.plot(lon0, lat0, marker='*', color='black', markersize=10)

# --------------------------
# (2,1) Time series: first two points
# --------------------------

for i in range(2):
    ax = axes[1, i]
    time_np = ts1[i].indexes['time'].to_datetimeindex()
    ax.plot(time_np, ts1[i].values, label=f'REF1 {points[i]}', lw=1.5)


ax.set_title('Deseasoned time series (Points 1–2)')
ax.set_ylabel('TEMP anomaly')
ax.legend(fontsize=8)

# --------------------------
# (2,2) Time series: last two points
# --------------------------
ax = axes[1, 1]
for i in range(2, 4):
    ax.plot(time_np, ts2[i].values, ls='--', label=f'REF2 {points[i]}', lw=1.5)
    # ax.plot(ts1[i].time, ts1[i], label=f'REF1 {points[i]}', lw=1.5)
    # ax.plot(ts2[i].time, ts2[i], ls='--', label=f'REF2 {points[i]}', lw=1.5)

ax.set_title('Deseasoned time series (Points 3–4)')
ax.set_ylabel('TEMP anomaly')
ax.legend(fontsize=8)

# --------------------------
# (3,*) empty panels (reserved)
# --------------------------
axes[2, 0].axis('off')
axes[2, 1].axis('off')

# --------------------------
# Colorbars
# --------------------------
cbar1 = fig.colorbar(pcm1, ax=axes[0, 0], extend='max')
cbar1.set_label('RMSE')

cbar2 = fig.colorbar(pcm2, ax=axes[0, 1])
cbar2.set_label('Correlation')

plt.show()
```
