```python
import xarray as xr
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=DeprecationWarning)

# ============================================================
# Settings
# ============================================================

START_YEAR = 1850
# END_YEAR = 1869   # first 20 years only
END_YEAR = 2014   # first 20 years only
# N_FILES = 2       # NorESM CMIP files are usually 10-year chunks
N_FILES = 17       # NorESM CMIP files are usually 10-year chunks

BASE_DIR = (
    "/nird/datapeak/NS9034K/CMIP6/CMIP/NCC/"
    "NorESM2-LM/esm-hist/r1i1p1f1"
)

EMISSION_FILE = (
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/TipESM/cmor/esm-up2p0/v20251010/AERA/emissions/reference/"
    "emissions-cmip6_CO2_anthro_surface_175001-201512_"
    "fv_1.9x2.5_c20181011.nc"
)

AREACELLO_FILE = (
    "/nird/datapeak/NS9034K/CMIP6/CMIP/NCC/NorESM2-LM/"
    "esm-hist/r1i1p1f1/Ofx/areacello/gn/latest/"
    "areacello_Ofx_NorESM2-LM_esm-hist_r1i1p1f1_gn.nc"
)

MODEL = "NorESM2-LM"
EXP = "esm-hist"
MEMBER = "r1i1p1f1"

# If the emission variable is kg CO2 m-2 s-1, use "CO2".
# If it is already kg C m-2 s-1, change this to "C".
EMISSION_MASS_BASIS = "CO2"   # "CO2" or "C"


PATTERNS = {
    "tas": (
        f"{BASE_DIR}/Amon/tas/gn/latest/"
        f"tas_Amon_{MODEL}_{EXP}_{MEMBER}_gn_*.nc"
    ),
    "co3os": (
        f"{BASE_DIR}/Omon/co3os/gr/latest/"
        f"co3os_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
    "co3sataragos": (
        f"{BASE_DIR}/Omon/co3sataragos/gr/latest/"
        f"co3sataragos_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
    "co3satcalcos": (
        f"{BASE_DIR}/Omon/co3satcalcos/gr/latest/"
        f"co3satcalcos_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
    "o2": (
        f"{BASE_DIR}/Omon/o2/gr/latest/"
        f"o2_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
}


# ============================================================
# Helper functions
# ============================================================

def get_first_files(pattern, n_files=N_FILES):
    files = sorted(glob.glob(pattern))
    print(f"\nPattern: {pattern}")
    print(f"Number of files found: {len(files)}")

    if len(files) == 0:
        raise FileNotFoundError(f"No files found:\n{pattern}")

    files = files[:n_files]
    print("Using files:")
    for f in files:
        print("  ", f)

    return files


def open_first_years(pattern, n_files=N_FILES):
    files = get_first_files(pattern, n_files=n_files)

    ds = xr.open_mfdataset(
        files,
        combine="by_coords",
        use_cftime=True,
        data_vars="minimal",
        coords="minimal",
        compat="override",
        chunks={"time": -1},
    )

    # Restrict to first 20 years
    ds = ds.sel(time=slice(f"{START_YEAR}-01-01", f"{END_YEAR}-12-31"))
    return ds


def find_lat(ds):
    for name in ["lat", "latitude", "nav_lat"]:
        if name in ds:
            return ds[name]
    raise KeyError("Could not find latitude variable.")


def find_lon(ds):
    for name in ["lon", "longitude", "nav_lon"]:
        if name in ds:
            return ds[name]
    raise KeyError("Could not find longitude variable.")


def get_spatial_dims(da):
    spatial_dims = [
        d for d in da.dims
        if d not in ["time", "lev", "olevel", "depth", "deptht", "bnds", "vertices"]
    ]
    return spatial_dims


def global_weighted_mean(da, ds):
    lat = find_lat(ds)
    spatial_dims = get_spatial_dims(da)

    weights = np.cos(np.deg2rad(lat))
    weights = weights.where(np.isfinite(da.isel(time=0)), 0.0).fillna(0.0)

    out = da.weighted(weights).mean(dim=spatial_dims, skipna=True).compute()
    return out


def annual_mean(da):
    out = da.groupby("time.year").mean("time")
    return out


def edges_from_centers_1d(x, is_lat=False):
    x = np.asarray(x, dtype=float)

    edges = np.empty(x.size + 1)
    edges[1:-1] = 0.5 * (x[:-1] + x[1:])
    edges[0] = x[0] - 0.5 * (x[1] - x[0])
    edges[-1] = x[-1] + 0.5 * (x[-1] - x[-2])

    if is_lat:
        edges = np.clip(edges, -90.0, 90.0)

    return edges


def approximate_area_from_latlon(ds, spatial_dims):
    """
    Approximate grid-cell area from latitude/longitude.
    This is fine for a quick gr-grid check.
    For final O2 inventory, using areacello/volcello is preferable.
    """
    radius = 6.371e6

    lat = find_lat(ds)
    lon = find_lon(ds)

    # Case 1: 1D lat/lon
    if lat.ndim == 1 and lon.ndim == 1:
        lat_dim = lat.dims[0]
        lon_dim = lon.dims[0]

        lat_edges = np.deg2rad(edges_from_centers_1d(lat.values, is_lat=True))
        lon_edges = np.deg2rad(edges_from_centers_1d(lon.values, is_lat=False))

        dlon = np.diff(lon_edges)
        area_lat = radius**2 * np.diff(np.sin(lat_edges))

        area = area_lat[:, None] * dlon[None, :]

        return xr.DataArray(
            area,
            dims=(lat_dim, lon_dim),
            coords={lat_dim: lat, lon_dim: lon},
            name="cell_area_approx",
            attrs={"units": "m2"},
        )

    # Case 2: 2D lat/lon on regular gr grid, usually dims like j,i
    if lat.ndim == 2 and lon.ndim == 2:
        ydim, xdim = lat.dims

        lat_1d = lat.isel({xdim: 0}).values
        lon_1d = lon.isel({ydim: 0}).values

        lat_edges = np.deg2rad(edges_from_centers_1d(lat_1d, is_lat=True))
        lon_edges = np.deg2rad(edges_from_centers_1d(lon_1d, is_lat=False))

        dlon = np.diff(lon_edges)
        area_lat = radius**2 * np.diff(np.sin(lat_edges))

        area = area_lat[:, None] * dlon[None, :]

        return xr.DataArray(
            area,
            dims=(ydim, xdim),
            coords={ydim: ds[ydim], xdim: ds[xdim]},
            name="cell_area_approx",
            attrs={"units": "m2"},
        )

    raise ValueError("Unsupported latitude/longitude shape.")


def find_vertical_dim(da):
    for name in ["lev", "olevel", "depth", "deptht"]:
        if name in da.dims:
            return name

    spatial_dims = get_spatial_dims(da)
    candidates = [d for d in da.dims if d not in ["time"] + spatial_dims]
    if len(candidates) == 1:
        return candidates[0]

    raise ValueError(f"Could not identify vertical dimension from dims: {da.dims}")


def get_vertical_bounds(ds, zdim):
    """
    Return vertical bounds in meters.
    This version avoids xr.apply_ufunc and loads only the vertical bounds.
    """
    candidates = [
        f"{zdim}_bnds",
        f"{zdim}_bounds",
        "lev_bnds",
        "olevel_bnds",
        "depth_bnds",
        "deptht_bnds",
    ]

    for name in candidates:
        if name in ds:
            zb = ds[name]

            # Identify the bounds dimension
            bdim_candidates = [d for d in zb.dims if d != zdim]
            if len(bdim_candidates) != 1:
                raise ValueError(f"Could not identify bounds dim for {name}: {zb.dims}")

            bdim = bdim_candidates[0]
            if bdim != "bnds":
                zb = zb.rename({bdim: "bnds"})

            zb = zb.load()

            units = zb.attrs.get("units", ds[zdim].attrs.get("units", "")).lower()
            if "cm" in units:
                zb = zb / 100.0
                zb.attrs["units"] = "m"

            return zb

    # If no bounds exist, estimate from vertical centers
    z = ds[zdim].values.astype(float)

    units = ds[zdim].attrs.get("units", "").lower()
    if "cm" in units:
        z = z / 100.0

    edges = edges_from_centers_1d(z, is_lat=False)
    bounds = np.vstack([edges[:-1], edges[1:]]).T

    zb = xr.DataArray(
        bounds,
        dims=(zdim, "bnds"),
        coords={zdim: ds[zdim]},
        name=f"{zdim}_bnds_estimated",
        attrs={"units": "m"},
    )

    return zb


def layer_thickness_0_to_zmax(ds, zdim, zmax=200.0):
    """
    Calculate partial layer thickness from surface to zmax.
    No apply_ufunc, so this works safely with dask-backed datasets.
    """
    zb = get_vertical_bounds(ds, zdim)

    zlo = zb.min("bnds")
    zhi = zb.max("bnds")

    upper = xr.where(zhi > zmax, zmax, zhi)
    lower = xr.where(zlo < 0.0, 0.0, zlo)

    thickness = upper - lower
    thickness = xr.where(thickness > 0.0, thickness, 0.0)

    thickness.name = f"layer_thickness_0_{int(zmax)}m"
    thickness.attrs["units"] = "m"

    return thickness

def load_areacello_or_fallback(ds_o2, spatial_dims):
    """
    Load areacello if its dimensions match the O2 grid.
    If not, fall back to approximate area from lat/lon.
    """
    try:
        ds_area = xr.open_dataset(AREACELLO_FILE)
        area = ds_area["areacello"].load()

        print("\nareacello info:")
        print(f"  file dims : {area.dims}")
        print(f"  file shape: {area.shape}")
        print(f"  O2 spatial dims : {spatial_dims}")
        print(f"  O2 spatial shape: {[ds_o2.sizes[d] for d in spatial_dims]}")

        # Direct match
        if set(area.dims) == set(spatial_dims):
            area = area.transpose(*spatial_dims)
            ds_area.close()
            print("Using areacello directly.")
            return area

        # Same order/shape but different dim names
        if len(area.dims) == len(spatial_dims):
            area_shape = [area.sizes[d] for d in area.dims]
            o2_shape = [ds_o2.sizes[d] for d in spatial_dims]

            if area_shape == o2_shape:
                rename_dict = {old: new for old, new in zip(area.dims, spatial_dims)}
                area = area.rename(rename_dict)
                area = area.transpose(*spatial_dims)
                ds_area.close()
                print("Using areacello after renaming dimensions.")
                return area

        ds_area.close()

        print(
            "WARNING: areacello grid does not match O2 grid. "
            "Falling back to approximate lat/lon area."
        )
        return approximate_area_from_latlon(ds_o2, spatial_dims)

    except Exception as e:
        print(f"WARNING: Could not use areacello: {e}")
        print("Falling back to approximate lat/lon area.")
        return approximate_area_from_latlon(ds_o2, spatial_dims)


# ============================================================
# Metric calculations
# ============================================================

def compute_tas_global_mean():
    ds = open_first_years(PATTERNS["tas"])
    tas = ds["tas"]

    tas_gm = global_weighted_mean(tas, ds)
    tas_ann = annual_mean(tas_gm)
    tas_ann.name = "global_mean_tas"

    ds.close()
    return tas_ann


def compute_omega_global_mean(co3sat_var):
    ds_co3 = open_first_years(PATTERNS["co3os"])
    ds_sat = open_first_years(PATTERNS[co3sat_var])

    co3 = ds_co3["co3os"]
    co3sat = ds_sat[co3sat_var]

    omega = co3 / co3sat

    if co3sat_var == "co3sataragos":
        omega.name = "omega_arag"
        omega.attrs["long_name"] = "Surface aragonite saturation state"
    elif co3sat_var == "co3satcalcos":
        omega.name = "omega_calc"
        omega.attrs["long_name"] = "Surface calcite saturation state"

    omega.attrs["units"] = "1"

    omega_gm = global_weighted_mean(omega, ds_co3)
    omega_ann = annual_mean(omega_gm)
    omega_ann.name = f"global_mean_{omega.name}"

    ds_co3.close()
    ds_sat.close()

    return omega_ann


def compute_o2_inventory_0_200m():
    ds = open_first_years(PATTERNS["o2"])
    o2 = ds["o2"]

    zdim = find_vertical_dim(o2)
    spatial_dims = get_spatial_dims(o2)

    print(f"O2 dims: {o2.dims}")
    print(f"Detected vertical dim: {zdim}")
    print(f"Detected spatial dims: {spatial_dims}")
    print(f"O2 units: {o2.attrs.get('units', 'unknown')}")

    thickness = layer_thickness_0_to_zmax(ds, zdim, zmax=200.0)
    area = load_areacello_or_fallback(ds, spatial_dims)

    # Unit conversion to mol m-3 if needed
    o2_units = o2.attrs.get("units", "").lower()
    if "mmol" in o2_units:
        print("Converting O2 from mmol m-3 to mol m-3")
        o2_for_inventory = o2 * 1e-3
    elif "umol" in o2_units or "µmol" in o2_units:
        print("Converting O2 from umol m-3 to mol m-3")
        o2_for_inventory = o2 * 1e-6
    else:
        print("Assuming O2 is already in mol m-3")
        o2_for_inventory = o2

    # Mask area where uppermost O2 is missing
    surface_mask = np.isfinite(o2_for_inventory.isel(time=0, **{zdim: 0}))
    area = area.where(surface_mask, 0.0).fillna(0.0)

    # Inventory = concentration × thickness × area
    inventory = (
        o2_for_inventory * thickness * area
    ).sum(
        dim=[zdim] + spatial_dims,
        skipna=True,
    ).compute()

    inventory.name = "global_o2_inventory_0_200m"
    inventory.attrs["units"] = "mol"

    inv_ann = annual_mean(inventory)
    inv_ann.name = "global_o2_inventory_0_200m_annual"

    ds.close()
    return inv_ann


# ============================================================
# Emission calculation
# ============================================================

def guess_emission_var(ds):
    preferred = [
        "CO2",
        "co2",
        "CO2_em_anthro",
        "co2_em_anthro",
        "emiss_anthro",
        "emissions",
    ]

    for name in preferred:
        if name in ds.data_vars:
            return name

    # fallback: choose first variable with time and lat/lon-like dims
    for name, da in ds.data_vars.items():
        dims = set(da.dims)
        if "time" in dims and any(d in dims for d in ["lat", "latitude"]) and any(d in dims for d in ["lon", "longitude"]):
            return name

    print("Available variables:")
    for name in ds.data_vars:
        print("  ", name, ds[name].dims, ds[name].attrs.get("units", ""))

    raise ValueError("Could not guess emission variable. Please set it manually.")


def compute_emission_annual_and_cumulative():
    ds = xr.open_dataset(EMISSION_FILE, use_cftime=True)

    emis = ds["CO2_flux"]
    print(f"Emission variable: CO2_flux")
    print(f"Emission dims: {emis.dims}")
    print(f"Emission units: {emis.attrs.get('units', 'unknown')}")
    print(f"Emission long_name: {emis.attrs.get('long_name', '')}")

    emis = emis.sel(time=slice(f"{START_YEAR}-01-01", f"{END_YEAR}-12-31"))

    lat = ds["lat"]
    lon = ds["lon"]

    # Grid-cell area for regular lat-lon grid
    area = approximate_area_from_latlon(ds, spatial_dims=["lat", "lon"])

    # monthly mean flux -> annual mean flux
    emis_ann_flux = emis.groupby("time.year").mean("time")

    seconds_per_year = 365.0 * 24.0 * 3600.0

    # kg CO2 m-2 s-1 * m2 * s yr-1 = kg CO2 yr-1
    emis_global_kgCO2_per_year = (
        emis_ann_flux * area * seconds_per_year
    ).sum(dim=["lat", "lon"], skipna=True).compute()

    # kg CO2 yr-1 -> Pg C yr-1
    emis_global_pgC_per_year = (
        emis_global_kgCO2_per_year / 1e12 * (12.0 / 44.0)
    )

    emis_global_pgC_per_year.name = "surface_anthro_CO2_emission_PgC_yr"

    emis_cum_pgC = emis_global_pgC_per_year.cumsum("year")
    emis_cum_pgC.name = "cumulative_surface_anthro_CO2_emission_PgC"

    ds.close()

    return emis_global_pgC_per_year, emis_cum_pgC


# ============================================================
# Run calculations
# ============================================================

tas_ann = compute_tas_global_mean()
omega_arag_ann = compute_omega_global_mean("co3sataragos")
omega_calc_ann = compute_omega_global_mean("co3satcalcos")
o2_inv_ann = compute_o2_inventory_0_200m()
emis_ann, emis_cum = compute_emission_annual_and_cumulative()

# ============================================================
# Save 1D annual time series
# ============================================================

OUT_DIR = (
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/TipESM/cmor/esm-up2p0/v20251010/AERA/input"
)

os.makedirs(OUT_DIR, exist_ok=True)

OUT_PREFIX = f"NorESM2-LM_esm-hist_r1i1p1f1_{START_YEAR}-{END_YEAR}"

# Make sure names are clear
tas_ann.name = "tas_global_mean_K"
omega_arag_ann.name = "omega_arag_global_mean"
omega_calc_ann.name = "omega_calc_global_mean"
o2_inv_ann.name = "o2_inventory_0_200m_mol"
emis_ann.name = "co2_surface_emission_PgC_yr"
emis_cum.name = "co2_surface_cumulative_emission_PgC"

# Combine into one Dataset
ds_ts = xr.Dataset(
    {
        "tas_global_mean_K": tas_ann,
        "omega_arag_global_mean": omega_arag_ann,
        "omega_calc_global_mean": omega_calc_ann,
        "o2_inventory_0_200m_mol": o2_inv_ann,
        "o2_inventory_0_200m_Pmol": o2_inv_ann / 1e15,
        "co2_surface_emission_PgC_yr": emis_ann,
        "co2_surface_cumulative_emission_PgC": emis_cum,
    }
)

# Add simple metadata
ds_ts["tas_global_mean_K"].attrs["units"] = "K"
ds_ts["omega_arag_global_mean"].attrs["units"] = "1"
ds_ts["omega_calc_global_mean"].attrs["units"] = "1"
ds_ts["o2_inventory_0_200m_mol"].attrs["units"] = "mol"
ds_ts["o2_inventory_0_200m_Pmol"].attrs["units"] = "Pmol"
ds_ts["co2_surface_emission_PgC_yr"].attrs["units"] = "Pg C yr-1"
ds_ts["co2_surface_cumulative_emission_PgC"].attrs["units"] = "Pg C"

ds_ts.attrs["model"] = "NorESM2-LM"
ds_ts.attrs["experiment"] = "esm-hist"
ds_ts.attrs["member"] = "r1i1p1f1"
ds_ts.attrs["period"] = f"{START_YEAR}-{END_YEAR}"
ds_ts.attrs["note"] = (
    "Annual global mean/sum time series for AERA metric testing. "
    "CO2 emissions are based on surface anthropogenic CO2_flux only."
)

# Save NetCDF
nc_out = os.path.join(OUT_DIR, f"{OUT_PREFIX}_annual_global_timeseries.nc")
ds_ts.to_netcdf(nc_out)

# Save wide-format CSV: one row per year, one column per variable
csv_out = os.path.join(OUT_DIR, f"{OUT_PREFIX}_annual_global_timeseries.csv")
df_ts = ds_ts.to_dataframe().reset_index()
df_ts.to_csv(csv_out, index=False)

print(f"Saved NetCDF: {nc_out}")
print(f"Saved CSV   : {csv_out}")

print(df_ts.head())

# ============================================================
# Plot 1: time series with cumulative emissions
# ============================================================

fig, axes = plt.subplots(4, 1, figsize=(10, 12), sharex=True)

plot_items = [
    (tas_ann, "Global mean TAS", "K"),
    (omega_arag_ann, r"Global mean surface $\Omega_{arag}$", "1"),
    (omega_calc_ann, r"Global mean surface $\Omega_{calc}$", "1"),
    (o2_inv_ann / 1e15, "Global O$_2$ inventory 0–200 m", "Pmol"),
]

for ax, (da, title, ylabel) in zip(axes, plot_items):
    ax.plot(da["year"], da, lw=2, label=title)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.grid(alpha=0.3)

    ax2 = ax.twinx()
    ax2.plot(emis_cum["year"], emis_cum, lw=1.5, ls="--", color="k", alpha=0.6)
    ax2.set_ylabel("Cumulative surface CO$_2$ emission (Pg C)")

axes[-1].set_xlabel("Year")
fig.suptitle("esm-hist first 20 years: metrics and cumulative CO$_2$ emissions", y=0.995)
plt.tight_layout()
plt.show()


# ============================================================
# Plot 2: variables against cumulative emissions
# ============================================================

fig, axes = plt.subplots(2, 2, figsize=(11, 8))
axes = axes.ravel()

plot_items = [
    (tas_ann, "Global mean TAS", "K"),
    (omega_arag_ann, r"Global mean surface $\Omega_{arag}$", "1"),
    (omega_calc_ann, r"Global mean surface $\Omega_{calc}$", "1"),
    (o2_inv_ann / 1e15, "Global O$_2$ inventory 0–200 m", "Pmol"),
]

for ax, (da, title, ylabel) in zip(axes, plot_items):
    common_years = np.intersect1d(da["year"].values, emis_cum["year"].values)

    x = emis_cum.sel(year=common_years)
    y = da.sel(year=common_years)

    ax.plot(x, y, marker="o", lw=2)
    ax.set_xlabel("Cumulative emission since 1850 (Pg C)")
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import xarray as xr
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore", category=DeprecationWarning)

# ============================================================
# Settings
# ============================================================

START_YEAR = 1850
# END_YEAR = 1869   # first 20 years only
END_YEAR = 2014   # first 20 years only
# N_FILES = 2       # NorESM CMIP files are usually 10-year chunks
N_FILES = 17       # NorESM CMIP files are usually 10-year chunks

BASE_DIR = (
    "/nird/datapeak/NS9034K/CMIP6/CMIP/NCC/"
    "NorESM2-LM/esm-hist/r1i1p1f1"
)

EMISSION_FILE = (
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/TipESM/cmor/esm-up2p0/v20251010/AERA/emissions/reference/"
    "emissions-cmip6_CO2_anthro_surface_175001-201512_"
    "fv_1.9x2.5_c20181011.nc"
)

AREACELLO_FILE = (
    "/nird/datapeak/NS9034K/CMIP6/CMIP/NCC/NorESM2-LM/"
    "esm-hist/r1i1p1f1/Ofx/areacello/gn/latest/"
    "areacello_Ofx_NorESM2-LM_esm-hist_r1i1p1f1_gn.nc"
)

MODEL = "NorESM2-LM"
EXP = "esm-hist"
MEMBER = "r1i1p1f1"

# If the emission variable is kg CO2 m-2 s-1, use "CO2".
# If it is already kg C m-2 s-1, change this to "C".
EMISSION_MASS_BASIS = "CO2"   # "CO2" or "C"


PATTERNS = {
    "tas": (
        f"{BASE_DIR}/Amon/tas/gn/latest/"
        f"tas_Amon_{MODEL}_{EXP}_{MEMBER}_gn_*.nc"
    ),
    "co3os": (
        f"{BASE_DIR}/Omon/co3os/gr/latest/"
        f"co3os_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
    "co3sataragos": (
        f"{BASE_DIR}/Omon/co3sataragos/gr/latest/"
        f"co3sataragos_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
    "co3satcalcos": (
        f"{BASE_DIR}/Omon/co3satcalcos/gr/latest/"
        f"co3satcalcos_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
    "o2": (
        f"{BASE_DIR}/Omon/o2/gr/latest/"
        f"o2_Omon_{MODEL}_{EXP}_{MEMBER}_gr_*.nc"
    ),
}


# ============================================================
# Helper functions
# ============================================================

def get_first_files(pattern, n_files=N_FILES):
    files = sorted(glob.glob(pattern))
    print(f"\nPattern: {pattern}")
    print(f"Number of files found: {len(files)}")

    if len(files) == 0:
        raise FileNotFoundError(f"No files found:\n{pattern}")

    files = files[:n_files]
    print("Using files:")
    for f in files:
        print("  ", f)

    return files


def open_first_years(pattern, n_files=N_FILES):
    files = get_first_files(pattern, n_files=n_files)

    ds = xr.open_mfdataset(
        files,
        combine="by_coords",
        use_cftime=True,
        data_vars="minimal",
        coords="minimal",
        compat="override",
        chunks={"time": 12},
    )

    # Restrict to first 20 years
    ds = ds.sel(time=slice(f"{START_YEAR}-01-01", f"{END_YEAR}-12-31"))
    return ds


def find_lat(ds):
    for name in ["lat", "latitude", "nav_lat"]:
        if name in ds:
            return ds[name]
    raise KeyError("Could not find latitude variable.")


def find_lon(ds):
    for name in ["lon", "longitude", "nav_lon"]:
        if name in ds:
            return ds[name]
    raise KeyError("Could not find longitude variable.")


def get_spatial_dims(da):
    spatial_dims = [
        d for d in da.dims
        if d not in ["time", "lev", "olevel", "depth", "deptht", "bnds", "vertices"]
    ]
    return spatial_dims


def global_weighted_mean(da, ds):
    lat = find_lat(ds)
    spatial_dims = get_spatial_dims(da)

    weights = np.cos(np.deg2rad(lat))
    weights = weights.where(np.isfinite(da.isel(time=0)), 0.0).fillna(0.0)

    out = da.weighted(weights).mean(dim=spatial_dims, skipna=True).compute()
    return out


def annual_mean(da):
    out = da.groupby("time.year").mean("time")
    return out


def edges_from_centers_1d(x, is_lat=False):
    x = np.asarray(x, dtype=float)

    edges = np.empty(x.size + 1)
    edges[1:-1] = 0.5 * (x[:-1] + x[1:])
    edges[0] = x[0] - 0.5 * (x[1] - x[0])
    edges[-1] = x[-1] + 0.5 * (x[-1] - x[-2])

    if is_lat:
        edges = np.clip(edges, -90.0, 90.0)

    return edges


def approximate_area_from_latlon(ds, spatial_dims):
    """
    Approximate grid-cell area from latitude/longitude.
    This is fine for a quick gr-grid check.
    For final O2 inventory, using areacello/volcello is preferable.
    """
    radius = 6.371e6

    lat = find_lat(ds)
    lon = find_lon(ds)

    # Case 1: 1D lat/lon
    if lat.ndim == 1 and lon.ndim == 1:
        lat_dim = lat.dims[0]
        lon_dim = lon.dims[0]

        lat_edges = np.deg2rad(edges_from_centers_1d(lat.values, is_lat=True))
        lon_edges = np.deg2rad(edges_from_centers_1d(lon.values, is_lat=False))

        dlon = np.diff(lon_edges)
        area_lat = radius**2 * np.diff(np.sin(lat_edges))

        area = area_lat[:, None] * dlon[None, :]

        return xr.DataArray(
            area,
            dims=(lat_dim, lon_dim),
            coords={lat_dim: lat, lon_dim: lon},
            name="cell_area_approx",
            attrs={"units": "m2"},
        )

    # Case 2: 2D lat/lon on regular gr grid, usually dims like j,i
    if lat.ndim == 2 and lon.ndim == 2:
        ydim, xdim = lat.dims

        lat_1d = lat.isel({xdim: 0}).values
        lon_1d = lon.isel({ydim: 0}).values

        lat_edges = np.deg2rad(edges_from_centers_1d(lat_1d, is_lat=True))
        lon_edges = np.deg2rad(edges_from_centers_1d(lon_1d, is_lat=False))

        dlon = np.diff(lon_edges)
        area_lat = radius**2 * np.diff(np.sin(lat_edges))

        area = area_lat[:, None] * dlon[None, :]

        return xr.DataArray(
            area,
            dims=(ydim, xdim),
            coords={ydim: ds[ydim], xdim: ds[xdim]},
            name="cell_area_approx",
            attrs={"units": "m2"},
        )

    raise ValueError("Unsupported latitude/longitude shape.")


def find_vertical_dim(da):
    for name in ["lev", "olevel", "depth", "deptht"]:
        if name in da.dims:
            return name

    spatial_dims = get_spatial_dims(da)
    candidates = [d for d in da.dims if d not in ["time"] + spatial_dims]
    if len(candidates) == 1:
        return candidates[0]

    raise ValueError(f"Could not identify vertical dimension from dims: {da.dims}")


def get_vertical_bounds(ds, zdim):
    """
    Return vertical bounds in meters.
    This version avoids xr.apply_ufunc and loads only the vertical bounds.
    """
    candidates = [
        f"{zdim}_bnds",
        f"{zdim}_bounds",
        "lev_bnds",
        "olevel_bnds",
        "depth_bnds",
        "deptht_bnds",
    ]

    for name in candidates:
        if name in ds:
            zb = ds[name]

            # Identify the bounds dimension
            bdim_candidates = [d for d in zb.dims if d != zdim]
            if len(bdim_candidates) != 1:
                raise ValueError(f"Could not identify bounds dim for {name}: {zb.dims}")

            bdim = bdim_candidates[0]
            if bdim != "bnds":
                zb = zb.rename({bdim: "bnds"})

            zb = zb.load()

            units = zb.attrs.get("units", ds[zdim].attrs.get("units", "")).lower()
            if "cm" in units:
                zb = zb / 100.0
                zb.attrs["units"] = "m"

            return zb

    # If no bounds exist, estimate from vertical centers
    z = ds[zdim].values.astype(float)

    units = ds[zdim].attrs.get("units", "").lower()
    if "cm" in units:
        z = z / 100.0

    edges = edges_from_centers_1d(z, is_lat=False)
    bounds = np.vstack([edges[:-1], edges[1:]]).T

    zb = xr.DataArray(
        bounds,
        dims=(zdim, "bnds"),
        coords={zdim: ds[zdim]},
        name=f"{zdim}_bnds_estimated",
        attrs={"units": "m"},
    )

    return zb


def layer_thickness_0_to_zmax(ds, zdim, zmax=200.0):
    """
    Calculate partial layer thickness from surface to zmax.
    No apply_ufunc, so this works safely with dask-backed datasets.
    """
    zb = get_vertical_bounds(ds, zdim)

    zlo = zb.min("bnds")
    zhi = zb.max("bnds")

    upper = xr.where(zhi > zmax, zmax, zhi)
    lower = xr.where(zlo < 0.0, 0.0, zlo)

    thickness = upper - lower
    thickness = xr.where(thickness > 0.0, thickness, 0.0)

    thickness.name = f"layer_thickness_0_{int(zmax)}m"
    thickness.attrs["units"] = "m"

    return thickness

def layer_thickness_between(ds, zdim, zmin, zmax):
    """
    Calculate partial layer thickness between zmin and zmax.
    Example: zmin=100, zmax=200 gives thickness contribution only
    from the 100-200 m depth range.
    """
    zb = get_vertical_bounds(ds, zdim)

    zlo = zb.min("bnds")
    zhi = zb.max("bnds")

    upper = xr.where(zhi > zmax, zmax, zhi)
    lower = xr.where(zlo < zmin, zmin, zlo)

    thickness = upper - lower
    thickness = xr.where(thickness > 0.0, thickness, 0.0)

    thickness.name = f"layer_thickness_{int(zmin)}_{int(zmax)}m"
    thickness.attrs["units"] = "m"

    return thickness

def load_areacello_or_fallback(ds_o2, spatial_dims):
    """
    Load areacello if its dimensions match the O2 grid.
    If not, fall back to approximate area from lat/lon.
    """
    try:
        ds_area = xr.open_dataset(AREACELLO_FILE)
        area = ds_area["areacello"].load()

        print("\nareacello info:")
        print(f"  file dims : {area.dims}")
        print(f"  file shape: {area.shape}")
        print(f"  O2 spatial dims : {spatial_dims}")
        print(f"  O2 spatial shape: {[ds_o2.sizes[d] for d in spatial_dims]}")

        # Direct match
        if set(area.dims) == set(spatial_dims):
            area = area.transpose(*spatial_dims)
            ds_area.close()
            print("Using areacello directly.")
            return area

        # Same order/shape but different dim names
        if len(area.dims) == len(spatial_dims):
            area_shape = [area.sizes[d] for d in area.dims]
            o2_shape = [ds_o2.sizes[d] for d in spatial_dims]

            if area_shape == o2_shape:
                rename_dict = {old: new for old, new in zip(area.dims, spatial_dims)}
                area = area.rename(rename_dict)
                area = area.transpose(*spatial_dims)
                ds_area.close()
                print("Using areacello after renaming dimensions.")
                return area

        ds_area.close()

        print(
            "WARNING: areacello grid does not match O2 grid. "
            "Falling back to approximate lat/lon area."
        )
        return approximate_area_from_latlon(ds_o2, spatial_dims)

    except Exception as e:
        print(f"WARNING: Could not use areacello: {e}")
        print("Falling back to approximate lat/lon area.")
        return approximate_area_from_latlon(ds_o2, spatial_dims)


# ============================================================
# Metric calculations
# ============================================================

def compute_tas_global_mean():
    ds = open_first_years(PATTERNS["tas"])
    tas = ds["tas"]

    tas_gm = global_weighted_mean(tas, ds)
    tas_ann = annual_mean(tas_gm)
    tas_ann.name = "global_mean_tas"

    ds.close()
    return tas_ann


def compute_omega_global_mean(co3sat_var):
    ds_co3 = open_first_years(PATTERNS["co3os"])
    ds_sat = open_first_years(PATTERNS[co3sat_var])

    co3 = ds_co3["co3os"]
    co3sat = ds_sat[co3sat_var]

    omega = co3 / co3sat

    if co3sat_var == "co3sataragos":
        omega.name = "omega_arag"
        omega.attrs["long_name"] = "Surface aragonite saturation state"
    elif co3sat_var == "co3satcalcos":
        omega.name = "omega_calc"
        omega.attrs["long_name"] = "Surface calcite saturation state"

    omega.attrs["units"] = "1"

    omega_gm = global_weighted_mean(omega, ds_co3)
    omega_ann = annual_mean(omega_gm)
    omega_ann.name = f"global_mean_{omega.name}"

    ds_co3.close()
    ds_sat.close()

    return omega_ann


def compute_o2_inventory_100m_layers_safe(zmax_total=1000.0, dz=100.0):
    """
    Memory-safe calculation of global O2 inventories for 100-m layers:
    0-100, 100-200, ..., 900-1000 m.

    This avoids creating a huge 5-D array:
    time x o2_layer x z x y x x.
    """
    import gc

    ds = open_first_years(PATTERNS["o2"])
    o2 = ds["o2"]

    zdim = find_vertical_dim(o2)
    spatial_dims = get_spatial_dims(o2)

    print(f"O2 dims: {o2.dims}")
    print(f"Detected vertical dim: {zdim}")
    print(f"Detected spatial dims: {spatial_dims}")
    print(f"O2 units: {o2.attrs.get('units', 'unknown')}")

    area = load_areacello_or_fallback(ds, spatial_dims)

    # Optional but useful to reduce memory
    area = area.astype("float32")

    # Unit conversion to mol m-3 if needed
    o2_units = o2.attrs.get("units", "").lower()

    if "mmol" in o2_units:
        print("Converting O2 from mmol m-3 to mol m-3")
        o2_for_inventory = (o2 * np.float32(1e-3)).astype("float32")
    elif "umol" in o2_units or "µmol" in o2_units:
        print("Converting O2 from umol m-3 to mol m-3")
        o2_for_inventory = (o2 * np.float32(1e-6)).astype("float32")
    else:
        print("Assuming O2 is already in mol m-3")
        o2_for_inventory = o2.astype("float32")

    layer_edges = np.arange(0.0, zmax_total + dz, dz)

    layer_ann_list = []
    layer_labels = []
    zmin_list = []
    zmax_list = []

    for zmin, zmax in zip(layer_edges[:-1], layer_edges[1:]):
        label = f"{int(zmin):04d}_{int(zmax):04d}m"
        print(f"\nComputing O2 inventory for layer {label}")

        thickness = layer_thickness_between(ds, zdim, zmin=zmin, zmax=zmax)
        thickness = thickness.astype("float32")

        # Use only vertical model levels that actually contribute to this layer
        active_idx = np.where(thickness.values > 0.0)[0]

        if len(active_idx) == 0:
            print(f"  No active vertical levels for {label}. Skipping.")
            continue

        print(f"  Active vertical indices: {active_idx}")

        o2_sel = o2_for_inventory.isel({zdim: active_idx})
        thickness_sel = thickness.isel({zdim: active_idx})

        # Monthly global inventory for this one layer only
        inv_month = (
            o2_sel * thickness_sel * area
        ).sum(
            dim=[zdim] + spatial_dims,
            skipna=True,
        )

        # Annual mean inventory
        inv_ann_layer = inv_month.groupby("time.year").mean("time").compute()

        inv_ann_layer.name = f"o2_inventory_{label}_mol"
        inv_ann_layer.attrs["units"] = "mol"
        inv_ann_layer.attrs["depth_layer"] = label

        layer_ann_list.append(inv_ann_layer)
        layer_labels.append(label)
        zmin_list.append(zmin)
        zmax_list.append(zmax)

        # Explicit cleanup
        del thickness, thickness_sel, o2_sel, inv_month, inv_ann_layer
        gc.collect()

    layer_coord = xr.DataArray(
        layer_labels,
        dims="o2_layer",
        name="o2_layer",
    )

    inv_ann = xr.concat(layer_ann_list, dim=layer_coord)
    inv_ann = inv_ann.transpose("year", "o2_layer")

    inv_ann = inv_ann.assign_coords(
        zmin_m=("o2_layer", zmin_list),
        zmax_m=("o2_layer", zmax_list),
    )

    inv_ann.name = "global_o2_inventory_layer_annual"
    inv_ann.attrs["units"] = "mol"
    inv_ann.attrs["description"] = "Global O2 inventory for 100-m depth layers"

    ds.close()
    gc.collect()

    return inv_ann

def o2_layers_to_dataset(o2_inv_layers_ann):
    """
    Convert O2 layer inventory DataArray with dimension (year, o2_layer)
    into a wide Dataset with one variable per layer.
    """
    ds_o2 = xr.Dataset()

    for label in o2_inv_layers_ann["o2_layer"].values:
        label = str(label)

        da_mol = (
            o2_inv_layers_ann
            .sel(o2_layer=label)
            .reset_coords(drop=True)
        )

        var_mol = f"o2_inventory_{label}_mol"
        var_pmol = f"o2_inventory_{label}_Pmol"

        ds_o2[var_mol] = da_mol
        ds_o2[var_mol].attrs["units"] = "mol"
        ds_o2[var_mol].attrs["depth_layer"] = label

        ds_o2[var_pmol] = da_mol / 1e15
        ds_o2[var_pmol].attrs["units"] = "Pmol"
        ds_o2[var_pmol].attrs["depth_layer"] = label

    return ds_o2

# ============================================================
# Emission calculation
# ============================================================

def guess_emission_var(ds):
    preferred = [
        "CO2",
        "co2",
        "CO2_em_anthro",
        "co2_em_anthro",
        "emiss_anthro",
        "emissions",
    ]

    for name in preferred:
        if name in ds.data_vars:
            return name

    # fallback: choose first variable with time and lat/lon-like dims
    for name, da in ds.data_vars.items():
        dims = set(da.dims)
        if "time" in dims and any(d in dims for d in ["lat", "latitude"]) and any(d in dims for d in ["lon", "longitude"]):
            return name

    print("Available variables:")
    for name in ds.data_vars:
        print("  ", name, ds[name].dims, ds[name].attrs.get("units", ""))

    raise ValueError("Could not guess emission variable. Please set it manually.")


def compute_emission_annual_and_cumulative():
    ds = xr.open_dataset(EMISSION_FILE, use_cftime=True)

    emis = ds["CO2_flux"]
    print(f"Emission variable: CO2_flux")
    print(f"Emission dims: {emis.dims}")
    print(f"Emission units: {emis.attrs.get('units', 'unknown')}")
    print(f"Emission long_name: {emis.attrs.get('long_name', '')}")

    emis = emis.sel(time=slice(f"{START_YEAR}-01-01", f"{END_YEAR}-12-31"))

    lat = ds["lat"]
    lon = ds["lon"]

    # Grid-cell area for regular lat-lon grid
    area = approximate_area_from_latlon(ds, spatial_dims=["lat", "lon"])

    # monthly mean flux -> annual mean flux
    emis_ann_flux = emis.groupby("time.year").mean("time")

    seconds_per_year = 365.0 * 24.0 * 3600.0

    # kg CO2 m-2 s-1 * m2 * s yr-1 = kg CO2 yr-1
    emis_global_kgCO2_per_year = (
        emis_ann_flux * area * seconds_per_year
    ).sum(dim=["lat", "lon"], skipna=True).compute()

    # kg CO2 yr-1 -> Pg C yr-1
    emis_global_pgC_per_year = (
        emis_global_kgCO2_per_year / 1e12 * (12.0 / 44.0)
    )

    emis_global_pgC_per_year.name = "surface_anthro_CO2_emission_PgC_yr"

    emis_cum_pgC = emis_global_pgC_per_year.cumsum("year")
    emis_cum_pgC.name = "cumulative_surface_anthro_CO2_emission_PgC"

    ds.close()

    return emis_global_pgC_per_year, emis_cum_pgC


# ============================================================
# Run calculations
# ============================================================

o2_inv_layers_ann = compute_o2_inventory_100m_layers_safe(
    zmax_total=1000.0,
    dz=100.0,
)

ds_o2_layers = o2_layers_to_dataset(o2_inv_layers_ann)
emis_ann, emis_cum = compute_emission_annual_and_cumulative()

# ============================================================
# Save 1D annual time series
# ============================================================

OUT_DIR = (
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/TipESM/cmor/esm-up2p0/v20251010/AERA/input"
)

os.makedirs(OUT_DIR, exist_ok=True)

OUT_PREFIX = f"NorESM2-LM_esm-hist_r1i1p1f1_{START_YEAR}-{END_YEAR}"

# Make sure names are clear
emis_ann.name = "co2_surface_emission_PgC_yr"
emis_cum.name = "co2_surface_cumulative_emission_PgC"

# Base Dataset
ds_base = xr.Dataset(
    {
        "co2_surface_emission_PgC_yr": emis_ann,
        "co2_surface_cumulative_emission_PgC": emis_cum,
    }
)

# Merge with O2 layer Dataset
ds_ts = xr.merge([ds_base, ds_o2_layers])

# Add simple metadata
ds_ts["co2_surface_emission_PgC_yr"].attrs["units"] = "Pg C yr-1"
ds_ts["co2_surface_cumulative_emission_PgC"].attrs["units"] = "Pg C"

ds_ts.attrs["model"] = "NorESM2-LM"
ds_ts.attrs["experiment"] = "esm-hist"
ds_ts.attrs["member"] = "r1i1p1f1"
ds_ts.attrs["period"] = f"{START_YEAR}-{END_YEAR}"
ds_ts.attrs["note"] = (
    "Annual global mean/sum time series for AERA metric testing. "
    "CO2 emissions are based on surface anthropogenic CO2_flux only. "
    "O2 inventories are calculated for 100-m layers from 0 to 1000 m."
)

# Save NetCDF
nc_out = os.path.join(OUT_DIR, f"{OUT_PREFIX}_annual_global_timeseries_o2_100m_layers.nc")
ds_ts.to_netcdf(nc_out)

# Save wide-format CSV
csv_out = os.path.join(OUT_DIR, f"{OUT_PREFIX}_annual_global_timeseries_o2_100m_layers.csv")
df_ts = ds_ts.to_dataframe().reset_index()
df_ts.to_csv(csv_out, index=False)

print(f"Saved NetCDF: {nc_out}")
print(f"Saved CSV   : {csv_out}")

print(df_ts.head())

# ============================================================
# Plot: O2 layer time series with cumulative emissions
# ============================================================

layer_labels = [str(x) for x in o2_inv_layers_ann["o2_layer"].values]

fig, axes = plt.subplots(5, 2, figsize=(13, 16), sharex=True)
axes = axes.ravel()

for ax, label in zip(axes, layer_labels):
    da = ds_o2_layers[f"o2_inventory_{label}_Pmol"]

    ax.plot(
        da["year"],
        da,
        lw=2,
        label=f"O2 inventory {label}",
    )

    ax.set_title(f"O$_2$ inventory {label}")
    ax.set_ylabel("Pmol")
    ax.grid(alpha=0.3)

    ax2 = ax.twinx()
    ax2.plot(
        emis_cum["year"],
        emis_cum,
        lw=1.2,
        ls="--",
        color="k",
        alpha=0.5,
    )
    ax2.set_ylabel("Cum. CO$_2$ (Pg C)")

for ax in axes[len(layer_labels):]:
    ax.axis("off")

axes[-1].set_xlabel("Year")
axes[-2].set_xlabel("Year")

fig.suptitle(
    "esm-hist: O$_2$ inventory by 100-m layers and cumulative CO$_2$ emissions",
    y=0.995,
)
plt.tight_layout()
plt.show()


# ============================================================
# Plot: O2 layer inventories against cumulative emissions
# ============================================================

fig, axes = plt.subplots(5, 2, figsize=(13, 16))
axes = axes.ravel()

for ax, label in zip(axes, layer_labels):
    da = ds_o2_layers[f"o2_inventory_{label}_Pmol"]

    common_years = np.intersect1d(
        da["year"].values,
        emis_cum["year"].values,
    )

    x = emis_cum.sel(year=common_years)
    y = da.sel(year=common_years)

    ax.plot(
        x,
        y,
        marker="o",
        ms=3,
        lw=1.5,
    )

    ax.set_xlabel("Cumulative emission since 1850 (Pg C)")
    ax.set_ylabel("O$_2$ inventory (Pmol)")
    ax.set_title(f"O$_2$ inventory {label}")
    ax.grid(alpha=0.3)

for ax in axes[len(layer_labels):]:
    ax.axis("off")

fig.suptitle(
    "esm-hist: O$_2$ inventory by 100-m layers vs cumulative CO$_2$ emissions",
    y=0.995,
)
plt.tight_layout()
plt.show()
```
