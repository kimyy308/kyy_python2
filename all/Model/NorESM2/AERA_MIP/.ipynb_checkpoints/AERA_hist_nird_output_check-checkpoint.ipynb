```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from pathlib import Path


# ============================================================
# Settings: must match the calculation script
# ============================================================
AERAid = "AERA_T_hist_nird"

FHIST_AERA_INPUT = Path(
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/"
    "TipESM/cmor/esm-up2p0/v20251010/AERA/input/references/"
    "NorESM2-LM_esm-hist_r1i1p1f1_1850-2014_AERA_full_input.csv"
)

y0 = 1850
y2 = 2014
nyr = 5

AERA_DIR = Path(
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/"
    "TipESM/cmor/esm-up2p0/v20251010/AERA"
)

OUTPUT_DIR = AERA_DIR / "output" / AERAid

FOUT = OUTPUT_DIR / f"{AERAid}_ann_tas_co2_ems_{y2}.dat"
meta_file = OUTPUT_DIR / f"{AERAid}_meta_{y2}.nc"
AERA_outfile = OUTPUT_DIR / f"{AERAid}_{y2}.csv"

PLOT_DIR = OUTPUT_DIR / f"figures_check_{y2}"
PLOT_DIR.mkdir(parents=True, exist_ok=True)


# ============================================================
# Helper functions
# ============================================================
def check_exists(path):
    if not Path(path).exists():
        raise FileNotFoundError(f"File does not exist: {path}")


def savefig(name):
    out = PLOT_DIR / name
    plt.tight_layout()
    plt.savefig(out, dpi=200, bbox_inches="tight")
    plt.show()
    print(f"Saved: {out}")


# ============================================================
# Check required files
# ============================================================
check_exists(FHIST_AERA_INPUT)
check_exists(FOUT)
check_exists(AERA_outfile)


# ============================================================
# 1. Read input CSV used for AERA calculation
# ============================================================
df_input = pd.read_csv(FHIST_AERA_INPUT, index_col="year")
df_input.index = df_input.index.astype(int)
df_input = df_input.sort_index()

input_years = np.arange(y0, y2 + 1)

input_temp = df_input.loc[input_years, "temp"]
input_ff = df_input.loc[input_years, "ff_emission"]


# ============================================================
# 2. Read pickle output
# ============================================================
with open(FOUT, "rb") as f:
    year, ann_tas, ann_co2, ann_ems = pickle.load(f)

year = np.asarray(year)
ann_tas = np.asarray(ann_tas)
ann_ems = np.asarray(ann_ems)

valid_year = ~np.isnan(year)

year_valid = year[valid_year].astype(int)

tas_series = pd.Series(
    ann_tas[valid_year],
    index=year_valid,
    name="ann_tas",
)

ems_series = pd.Series(
    ann_ems[valid_year],
    index=year_valid,
    name="ann_ems",
)

tas_series = tas_series.dropna()
ems_series = ems_series.dropna()

next5_years = np.arange(y2 + 1, y2 + nyr + 1)


# ============================================================
# 3. Read AERA future emission CSV
# ============================================================
aera_future = pd.read_csv(AERA_outfile, index_col=0)
aera_future.index = aera_future.index.astype(int)

if "ff_emission" in aera_future.columns:
    future_ff = aera_future["ff_emission"].copy()
else:
    future_ff = aera_future.iloc[:, 0].copy()

future_ff.name = "ff_emission"
future_ff = pd.to_numeric(future_ff, errors="coerce").dropna()


# ============================================================
# 4. Read metadata CSVs if available
# ============================================================
timeseries_csv = Path(str(meta_file) + ".timeseries.csv")
scalar_csv = Path(str(meta_file) + ".scalar.csv")

meta_ts_run = None
meta_scalar = None

if timeseries_csv.exists():
    meta_ts = pd.read_csv(timeseries_csv, index_col=[0, 1])
    meta_ts.index.names = ["year_stocktake", "year"]

    meta_ts = meta_ts.apply(pd.to_numeric, errors="coerce")
    stocktake_years = meta_ts.index.get_level_values("year_stocktake").astype(int)

    if y2 in stocktake_years:
        meta_ts_run = meta_ts.loc[y2].copy()
        meta_ts_run.index = meta_ts_run.index.astype(int)

if scalar_csv.exists():
    meta_scalar_all = pd.read_csv(scalar_csv, index_col=0)
    meta_scalar_all.index = meta_scalar_all.index.astype(int)
    meta_scalar_all = meta_scalar_all.apply(pd.to_numeric, errors="coerce")

    if y2 in meta_scalar_all.index:
        meta_scalar = meta_scalar_all.loc[y2].copy()


# ============================================================
# Fig 1. Input temperature
# ============================================================
plt.figure(figsize=(9, 4.5))
plt.plot(input_temp.index, input_temp.values, lw=2, label="input temp")
plt.axvline(y2, ls="--", lw=1, label=f"stocktake year = {y2}")
plt.xlabel("Year")
plt.ylabel("Temperature (K)")
plt.title("Input GMSAT used for AERA")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_01_input_temperature_{y2}.png")


# ============================================================
# Fig 2. Input ff_emission and next 5-year stored emissions
# ============================================================
plt.figure(figsize=(9, 4.5))
plt.plot(
    ems_series.index,
    ems_series.values,
    lw=2,
    marker="o",
    label="stored ff_emission in pickle",
)
plt.axvline(y2, ls="--", lw=1, label=f"stocktake year = {y2}")
plt.axvspan(y2 + 1, y2 + nyr, alpha=0.15, label="next 5 years")
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("Fossil-fuel CO$_2$ emission stored in pickle")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_02_pickle_ff_emission_{y2}.png")


# ============================================================
# Fig 3. Full AERA future emission curve
# ============================================================
plt.figure(figsize=(9, 4.5))

plt.plot(
    input_ff.index,
    input_ff.values,
    lw=2,
    label="input historical ff_emission",
)

plt.plot(
    future_ff.index,
    future_ff.values,
    lw=2,
    label="AERA future ff_emission curve",
)

plt.scatter(
    next5_years,
    future_ff.loc[next5_years],
    s=50,
    zorder=5,
    label="next 5 years used",
)

plt.axvline(y2, ls="--", lw=1)
plt.axhline(0, lw=1)
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("AERA-calculated fossil-fuel emission curve")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_03_aera_future_ff_curve_{y2}.png")


# ============================================================
# Fig 4. Zoom around stocktake
# ============================================================
zoom_start = max(y0, y2 - 25)
zoom_end = y2 + 40

plt.figure(figsize=(9, 4.5))

plt.plot(
    input_ff.loc[zoom_start:y2].index,
    input_ff.loc[zoom_start:y2].values,
    lw=2,
    marker="o",
    label="input historical ff_emission",
)

plt.plot(
    future_ff.loc[y2 + 1:zoom_end].index,
    future_ff.loc[y2 + 1:zoom_end].values,
    lw=2,
    marker="o",
    label="AERA future curve",
)

plt.scatter(
    next5_years,
    future_ff.loc[next5_years],
    s=60,
    zorder=5,
    label="next 5 years used",
)

plt.axvline(y2, ls="--", lw=1)
plt.axhline(0, lw=1)
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("AERA emission curve near stocktake")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_04_zoom_ff_curve_{y2}.png")


# ============================================================
# Fig 5. Stored next-5-year emissions vs AERA_outfile
# ============================================================
pickle_next5 = ems_series.loc[next5_years]
csv_next5 = future_ff.loc[next5_years]
diff_next5 = pickle_next5 - csv_next5

plt.figure(figsize=(9, 4.5))
plt.plot(next5_years, pickle_next5, marker="o", lw=2, label="pickle ann_ems")
plt.plot(next5_years, csv_next5, marker="x", lw=2, label="AERA csv")
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("Check: next 5-year emissions in pickle vs AERA CSV")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_05_next5_pickle_vs_csv_{y2}.png")

print("\nNext 5-year emission check")
print(pd.DataFrame({
    "pickle_ann_ems": pickle_next5,
    "aera_csv": csv_next5,
    "difference": diff_next5,
}))


# ============================================================
# Fig 6. Cumulative fossil-fuel emissions
# ============================================================
combined_ff = pd.concat(
    [
        input_ff.loc[:y2],
        future_ff.loc[future_ff.index > y2],
    ]
).sort_index()

plt.figure(figsize=(9, 4.5))
plt.plot(combined_ff.index, combined_ff.cumsum(), lw=2)
plt.axvline(y2, ls="--", lw=1)
plt.xlabel("Year")
plt.ylabel("Cumulative ff emission (PgC)")
plt.title("Cumulative fossil-fuel CO$_2$ emissions")
plt.grid(alpha=0.3)
savefig(f"{AERAid}_check_06_cumulative_ff_emission_{y2}.png")


# ============================================================
# Fig 7. Metadata total_emission and ff_emission
# ============================================================
if meta_ts_run is not None:
    cols = [c for c in ["total_emission", "ff_emission"] if c in meta_ts_run.columns]

    if len(cols) > 0:
        plt.figure(figsize=(9, 4.5))

        for col in cols:
            plt.plot(
                meta_ts_run.index,
                meta_ts_run[col],
                lw=2,
                label=col,
            )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("Emission (PgC yr$^{-1}$)")
        plt.title("AERA metadata: total emission and ff emission")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_07_meta_total_vs_ff_emission_{y2}.png")


# ============================================================
# Fig 8. Metadata emission decomposition
# total_emission - ff_emission = lu_emission + non_co2_emission
# ============================================================
if meta_ts_run is not None:
    if ("total_emission" in meta_ts_run.columns) and ("ff_emission" in meta_ts_run.columns):
        residual = meta_ts_run["total_emission"] - meta_ts_run["ff_emission"]

        plt.figure(figsize=(9, 4.5))
        plt.plot(
            meta_ts_run.index,
            meta_ts_run["ff_emission"],
            lw=2,
            label="ff_emission",
        )
        plt.plot(
            meta_ts_run.index,
            residual,
            lw=2,
            label="lu + nonCO2 contribution",
        )
        plt.plot(
            meta_ts_run.index,
            meta_ts_run["total_emission"],
            lw=2,
            label="total_emission",
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("Emission (PgC yr$^{-1}$)")
        plt.title("AERA metadata: emission decomposition")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_08_meta_emission_decomposition_{y2}.png")


# ============================================================
# Fig 9. Anthropogenic temperature diagnostic
# ============================================================
if meta_ts_run is not None and "temp_anth" in meta_ts_run.columns:
    plt.figure(figsize=(9, 4.5))

    plt.plot(
        input_temp.index,
        input_temp.values,
        lw=1.5,
        alpha=0.6,
        label="input temp",
    )

    plt.plot(
        meta_ts_run.index,
        meta_ts_run["temp_anth"],
        lw=2.5,
        label="AERA temp_anth",
    )

    if meta_scalar is not None and "temp_target_abs" in meta_scalar.index:
        plt.axhline(
            meta_scalar["temp_target_abs"],
            ls="--",
            lw=1.5,
            label="absolute target temp",
        )

    plt.axvline(y2, ls="--", lw=1)
    plt.xlabel("Year")
    plt.ylabel("Temperature (K)")
    plt.title("AERA anthropogenic temperature diagnostic")
    plt.grid(alpha=0.3)
    plt.legend()
    savefig(f"{AERAid}_check_09_temp_anth_diagnostic_{y2}.png")


# ============================================================
# Fig 10. Relative anthropogenic temperature diagnostic
# ============================================================
if meta_ts_run is not None and "temp_anth_rel" in meta_ts_run.columns:
    plt.figure(figsize=(9, 4.5))

    plt.plot(
        meta_ts_run.index,
        meta_ts_run["temp_anth_rel"],
        lw=2,
        label="temp_anth_rel",
    )

    plt.axvline(y2, ls="--", lw=1)
    plt.axhline(0, lw=1)

    if meta_scalar is not None and "temp_target_abs" in meta_scalar.index:
        # temp_anth_rel is stored as:
        # temp_anth - temp_target_abs + temp_target_rel
        pass

    plt.xlabel("Year")
    plt.ylabel("Relative anthropogenic temperature (K)")
    plt.title("AERA relative anthropogenic temperature")
    plt.grid(alpha=0.3)
    plt.legend()
    savefig(f"{AERAid}_check_10_temp_anth_rel_{y2}.png")


# ============================================================
# Fig 11. Reconstruct cubic curve from metadata coefficients
# ============================================================
if meta_scalar is not None:
    required = ["ec_a", "ec_b", "ec_c", "ec_d", "ec_target_year_rel"]

    if all(k in meta_scalar.index for k in required):
        a = meta_scalar["ec_a"]
        b = meta_scalar["ec_b"]
        c = meta_scalar["ec_c"]
        d = meta_scalar["ec_d"]
        target_year_rel = int(meta_scalar["ec_target_year_rel"])

        t = np.arange(0, target_year_rel + 1)
        recon_year = y2 + t
        recon_ff = a * t**3 + b * t**2 + c * t + d

        plt.figure(figsize=(9, 4.5))

        plt.plot(
            recon_year,
            recon_ff,
            lw=2.5,
            label="reconstructed cubic curve",
        )

        plt.plot(
            future_ff.index,
            future_ff.values,
            lw=1.5,
            ls="--",
            label="saved AERA future ff_emission",
        )

        plt.scatter(
            next5_years,
            future_ff.loc[next5_years],
            s=60,
            zorder=5,
            label="next 5 years used",
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("ff emission (PgC yr$^{-1}$)")
        plt.title("Reconstructed AERA cubic emission curve")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_11_reconstructed_cubic_curve_{y2}.png")


# ============================================================
# Print scalar diagnostics
# ============================================================
if meta_scalar is not None:
    print("\nAERA scalar diagnostics")
    print("=" * 60)

    keys_to_show = [
        "temp_target_abs",
        "total_emission_budget",
        "ff_emission_budget",
        "reb",
        "ec_cost",
        "ec_reb_diff",
        "ec_overshoot_integral",
        "ec_overshoot",
        "ec_curvature",
        "ec_target_year_rel",
        "ec_a",
        "ec_b",
        "ec_c",
        "ec_d",
    ]

    for k in keys_to_show:
        if k in meta_scalar.index:
            print(f"{k:25s}: {meta_scalar[k]}")

print(f"\nAll figures saved in: {PLOT_DIR}")
```

```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from pathlib import Path


# ============================================================
# Settings: must match the OA calculation script
# ============================================================
AERAid = "AERA_OmegaA_hist_nird"

FHIST_AERA_INPUT = Path(
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/"
    "TipESM/cmor/esm-up2p0/v20251010/AERA/input/references/"
    "NorESM2-LM_esm-hist_r1i1p1f1_1850-2014_AERA_OmegaA_full_input.csv"
)

OA_COL = "OmegaA"
OA_TARGET_ABS = 2.75

y0 = 1850
y2 = 2014
nyr = 5

AERA_DIR = Path(
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/"
    "TipESM/cmor/esm-up2p0/v20251010/AERA"
)

OUTPUT_DIR = AERA_DIR / "output" / AERAid

FOUT = OUTPUT_DIR / f"{AERAid}_ann_OmegaA_co2_ems_{y2}.dat"
meta_file = OUTPUT_DIR / f"{AERAid}_meta_{y2}.nc"
AERA_outfile = OUTPUT_DIR / f"{AERAid}_{y2}.csv"

PLOT_DIR = OUTPUT_DIR / f"figures_check_{y2}"
PLOT_DIR.mkdir(parents=True, exist_ok=True)


# ============================================================
# Helper functions
# ============================================================
def check_exists(path):
    if not Path(path).exists():
        raise FileNotFoundError(f"File does not exist: {path}")


def savefig(name):
    out = PLOT_DIR / name
    plt.tight_layout()
    plt.savefig(out, dpi=200, bbox_inches="tight")
    plt.show()
    print(f"Saved: {out}")


def first_existing_column(df, candidates):
    for c in candidates:
        if c in df.columns:
            return c
    return None


def first_existing_index(series, candidates):
    for c in candidates:
        if c in series.index:
            return c
    return None


# ============================================================
# Check required files
# ============================================================
check_exists(FHIST_AERA_INPUT)
check_exists(FOUT)
check_exists(AERA_outfile)


# ============================================================
# 1. Read input CSV used for AERA-OA calculation
# ============================================================
df_input = pd.read_csv(FHIST_AERA_INPUT, index_col="year")
df_input.index = df_input.index.astype(int)
df_input = df_input.sort_index()

input_years = np.arange(y0, y2 + 1)

if OA_COL not in df_input.columns:
    raise ValueError(
        f"Input CSV does not contain '{OA_COL}'. "
        f"Available columns: {list(df_input.columns)}"
    )

if "ff_emission" not in df_input.columns:
    raise ValueError(
        f"Input CSV does not contain 'ff_emission'. "
        f"Available columns: {list(df_input.columns)}"
    )

input_omegaa = pd.to_numeric(df_input.loc[input_years, OA_COL], errors="coerce")
input_ff = pd.to_numeric(df_input.loc[input_years, "ff_emission"], errors="coerce")


# ============================================================
# 2. Read pickle output
# ============================================================
with open(FOUT, "rb") as f:
    year, ann_omegaa, ann_co2, ann_ems = pickle.load(f)

year = np.asarray(year)
ann_omegaa = np.asarray(ann_omegaa)
ann_ems = np.asarray(ann_ems)

valid_year = ~np.isnan(year)
year_valid = year[valid_year].astype(int)

omegaa_series = pd.Series(
    ann_omegaa[valid_year],
    index=year_valid,
    name="ann_OmegaA",
)

ems_series = pd.Series(
    ann_ems[valid_year],
    index=year_valid,
    name="ann_ems",
)

omegaa_series = omegaa_series.dropna()
ems_series = ems_series.dropna()

next5_years = np.arange(y2 + 1, y2 + nyr + 1)


# ============================================================
# 3. Read AERA-OA future emission CSV
# ============================================================
aera_future = pd.read_csv(AERA_outfile, index_col=0)
aera_future.index = aera_future.index.astype(int)

if "ff_emission" in aera_future.columns:
    future_ff = aera_future["ff_emission"].copy()
else:
    future_ff = aera_future.iloc[:, 0].copy()

future_ff.name = "ff_emission"
future_ff = pd.to_numeric(future_ff, errors="coerce").dropna()


# ============================================================
# 4. Read metadata CSVs if available
# ============================================================
timeseries_csv = Path(str(meta_file) + ".timeseries.csv")
scalar_csv = Path(str(meta_file) + ".scalar.csv")

meta_ts_run = None
meta_scalar = None

if timeseries_csv.exists():
    meta_ts = pd.read_csv(timeseries_csv, index_col=[0, 1])
    meta_ts.index.names = ["year_stocktake", "year"]

    meta_ts = meta_ts.apply(pd.to_numeric, errors="coerce")
    stocktake_years = meta_ts.index.get_level_values("year_stocktake").astype(int)

    if y2 in stocktake_years:
        meta_ts_run = meta_ts.loc[y2].copy()
        meta_ts_run.index = meta_ts_run.index.astype(int)

if scalar_csv.exists():
    meta_scalar_all = pd.read_csv(scalar_csv, index_col=0)
    meta_scalar_all.index = meta_scalar_all.index.astype(int)
    meta_scalar_all = meta_scalar_all.apply(pd.to_numeric, errors="coerce")

    if y2 in meta_scalar_all.index:
        meta_scalar = meta_scalar_all.loc[y2].copy()


# ============================================================
# Fig 1. Input OmegaA
# ============================================================
plt.figure(figsize=(9, 4.5))
plt.plot(input_omegaa.index, input_omegaa.values, lw=2, label="input OmegaA")
plt.axhline(OA_TARGET_ABS, ls="--", lw=1.5, label=f"OmegaA target = {OA_TARGET_ABS}")
plt.axvline(y2, ls="--", lw=1, label=f"stocktake year = {y2}")
plt.xlabel("Year")
plt.ylabel(r"$\Omega_{\mathrm{arag}}$")
plt.title("Input OmegaA used for AERA-OA")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_01_input_OmegaA_{y2}.png")


# ============================================================
# Fig 2. Input ff_emission and next 5-year stored emissions
# ============================================================
plt.figure(figsize=(9, 4.5))
plt.plot(
    ems_series.index,
    ems_series.values,
    lw=2,
    marker="o",
    label="stored ff_emission in pickle",
)
plt.axvline(y2, ls="--", lw=1, label=f"stocktake year = {y2}")
plt.axvspan(y2 + 1, y2 + nyr, alpha=0.15, label="next 5 years")
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("Fossil-fuel CO$_2$ emission stored in pickle")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_02_pickle_ff_emission_{y2}.png")


# ============================================================
# Fig 3. Full AERA-OA future emission curve
# ============================================================
plt.figure(figsize=(9, 4.5))

plt.plot(
    input_ff.index,
    input_ff.values,
    lw=2,
    label="input historical ff_emission",
)

plt.plot(
    future_ff.index,
    future_ff.values,
    lw=2,
    label="AERA-OA future ff_emission curve",
)

plt.scatter(
    next5_years,
    future_ff.loc[next5_years],
    s=50,
    zorder=5,
    label="next 5 years used",
)

plt.axvline(y2, ls="--", lw=1)
plt.axhline(0, lw=1)
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("AERA-OA-calculated fossil-fuel emission curve")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_03_aera_oa_future_ff_curve_{y2}.png")


# ============================================================
# Fig 4. Zoom around stocktake
# ============================================================
zoom_start = max(y0, y2 - 25)
zoom_end = y2 + 40

plt.figure(figsize=(9, 4.5))

plt.plot(
    input_ff.loc[zoom_start:y2].index,
    input_ff.loc[zoom_start:y2].values,
    lw=2,
    marker="o",
    label="input historical ff_emission",
)

plt.plot(
    future_ff.loc[y2 + 1:zoom_end].index,
    future_ff.loc[y2 + 1:zoom_end].values,
    lw=2,
    marker="o",
    label="AERA-OA future curve",
)

plt.scatter(
    next5_years,
    future_ff.loc[next5_years],
    s=60,
    zorder=5,
    label="next 5 years used",
)

plt.axvline(y2, ls="--", lw=1)
plt.axhline(0, lw=1)
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("AERA-OA emission curve near stocktake")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_04_zoom_ff_curve_{y2}.png")


# ============================================================
# Fig 5. Stored next-5-year emissions vs AERA_outfile
# ============================================================
pickle_next5 = ems_series.loc[next5_years]
csv_next5 = future_ff.loc[next5_years]
diff_next5 = pickle_next5 - csv_next5

plt.figure(figsize=(9, 4.5))
plt.plot(next5_years, pickle_next5, marker="o", lw=2, label="pickle ann_ems")
plt.plot(next5_years, csv_next5, marker="x", lw=2, label="AERA-OA csv")
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("Check: next 5-year emissions in pickle vs AERA-OA CSV")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_05_next5_pickle_vs_csv_{y2}.png")

print("\nNext 5-year emission check")
print(pd.DataFrame({
    "pickle_ann_ems": pickle_next5,
    "aera_oa_csv": csv_next5,
    "difference": diff_next5,
}))


# ============================================================
# Fig 6. Cumulative fossil-fuel emissions
# ============================================================
combined_ff = pd.concat(
    [
        input_ff.loc[:y2],
        future_ff.loc[future_ff.index > y2],
    ]
).sort_index()

plt.figure(figsize=(9, 4.5))
plt.plot(combined_ff.index, combined_ff.cumsum(), lw=2)
plt.axvline(y2, ls="--", lw=1)
plt.xlabel("Year")
plt.ylabel("Cumulative ff emission (PgC)")
plt.title("Cumulative fossil-fuel CO$_2$ emissions")
plt.grid(alpha=0.3)
savefig(f"{AERAid}_check_06_cumulative_ff_emission_{y2}.png")


# ============================================================
# Fig 7. Metadata total_emission and ff_emission
# ============================================================
if meta_ts_run is not None:
    cols = [c for c in ["total_emission", "ff_emission"] if c in meta_ts_run.columns]

    if len(cols) > 0:
        plt.figure(figsize=(9, 4.5))

        for col in cols:
            plt.plot(
                meta_ts_run.index,
                meta_ts_run[col],
                lw=2,
                label=col,
            )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("Emission (PgC yr$^{-1}$)")
        plt.title("AERA-OA metadata: total emission and ff emission")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_07_meta_total_vs_ff_emission_{y2}.png")


# ============================================================
# Fig 8. Metadata emission decomposition
# total_emission - ff_emission = lu_emission + non_co2_emission
# ============================================================
if meta_ts_run is not None:
    if ("total_emission" in meta_ts_run.columns) and ("ff_emission" in meta_ts_run.columns):
        residual = meta_ts_run["total_emission"] - meta_ts_run["ff_emission"]

        plt.figure(figsize=(9, 4.5))
        plt.plot(
            meta_ts_run.index,
            meta_ts_run["ff_emission"],
            lw=2,
            label="ff_emission",
        )
        plt.plot(
            meta_ts_run.index,
            residual,
            lw=2,
            label="lu + nonCO2 contribution",
        )
        plt.plot(
            meta_ts_run.index,
            meta_ts_run["total_emission"],
            lw=2,
            label="total_emission",
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("Emission (PgC yr$^{-1}$)")
        plt.title("AERA-OA metadata: emission decomposition")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_08_meta_emission_decomposition_{y2}.png")


# ============================================================
# Fig 9. OmegaA diagnostic from metadata
# Robust to different possible metadata column names.
# ============================================================
if meta_ts_run is not None:
    omegaa_anth_col = first_existing_column(
        meta_ts_run,
        [
            "arag_anth",
            "omegaa_anth",
            "OmegaA_anth",
            "omega_arag_anth",
            "temp_anth",       # fallback if code still uses old internal name
        ],
    )

    if omegaa_anth_col is not None:
        plt.figure(figsize=(9, 4.5))

        plt.plot(
            input_omegaa.index,
            input_omegaa.values,
            lw=1.5,
            alpha=0.6,
            label="input OmegaA",
        )

        plt.plot(
            meta_ts_run.index,
            meta_ts_run[omegaa_anth_col],
            lw=2.5,
            label=f"AERA-OA {omegaa_anth_col}",
        )

        target_key = None
        if meta_scalar is not None:
            target_key = first_existing_index(
                meta_scalar,
                [
                    "arag_target_abs",
                    "omegaa_target_abs",
                    "OmegaA_target_abs",
                    "omega_arag_target_abs",
                    "temp_target_abs",  # fallback
                ],
            )

        if target_key is not None:
            plt.axhline(
                meta_scalar[target_key],
                ls="--",
                lw=1.5,
                label=f"target from metadata ({target_key})",
            )
        else:
            plt.axhline(
                OA_TARGET_ABS,
                ls="--",
                lw=1.5,
                label=f"OmegaA target = {OA_TARGET_ABS}",
            )

        plt.axvline(y2, ls="--", lw=1)
        plt.xlabel("Year")
        plt.ylabel(r"$\Omega_{\mathrm{arag}}$")
        plt.title("AERA-OA OmegaA diagnostic")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_09_OmegaA_diagnostic_{y2}.png")


# ============================================================
# Fig 10. Relative / target-gap OmegaA diagnostic if available
# Robust to different possible metadata column names.
# ============================================================
if meta_ts_run is not None:
    omegaa_rel_col = first_existing_column(
        meta_ts_run,
        [
            "arag_anth_rel",
            "omegaa_anth_rel",
            "OmegaA_anth_rel",
            "omega_arag_anth_rel",
            "temp_anth_rel",   # fallback if code still uses old internal name
        ],
    )

    if omegaa_rel_col is not None:
        plt.figure(figsize=(9, 4.5))

        plt.plot(
            meta_ts_run.index,
            meta_ts_run[omegaa_rel_col],
            lw=2,
            label=omegaa_rel_col,
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)

        plt.xlabel("Year")
        plt.ylabel("Relative / target-gap OmegaA")
        plt.title("AERA-OA relative OmegaA diagnostic")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_10_OmegaA_relative_diagnostic_{y2}.png")


# ============================================================
# Fig 11. Reconstruct cubic curve from metadata coefficients
# ============================================================
if meta_scalar is not None:
    required = ["ec_a", "ec_b", "ec_c", "ec_d", "ec_target_year_rel"]

    if all(k in meta_scalar.index for k in required):
        a = meta_scalar["ec_a"]
        b = meta_scalar["ec_b"]
        c = meta_scalar["ec_c"]
        d = meta_scalar["ec_d"]
        target_year_rel = int(meta_scalar["ec_target_year_rel"])

        t = np.arange(0, target_year_rel + 1)
        recon_year = y2 + t
        recon_ff = a * t**3 + b * t**2 + c * t + d

        plt.figure(figsize=(9, 4.5))

        plt.plot(
            recon_year,
            recon_ff,
            lw=2.5,
            label="reconstructed cubic curve",
        )

        plt.plot(
            future_ff.index,
            future_ff.values,
            lw=1.5,
            ls="--",
            label="saved AERA-OA future ff_emission",
        )

        plt.scatter(
            next5_years,
            future_ff.loc[next5_years],
            s=60,
            zorder=5,
            label="next 5 years used",
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("ff emission (PgC yr$^{-1}$)")
        plt.title("Reconstructed AERA-OA cubic emission curve")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_11_reconstructed_cubic_curve_{y2}.png")


# ============================================================
# Print scalar diagnostics
# ============================================================
if meta_scalar is not None:
    print("\nAERA-OA scalar diagnostics")
    print("=" * 60)

    keys_to_show = [
        "arag_target_abs",
        "omegaa_target_abs",
        "OmegaA_target_abs",
        "omega_arag_target_abs",
        "temp_target_abs",
        "total_emission_budget",
        "ff_emission_budget",
        "reb",
        "ec_cost",
        "ec_reb_diff",
        "ec_overshoot_integral",
        "ec_overshoot",
        "ec_curvature",
        "ec_target_year_rel",
        "ec_a",
        "ec_b",
        "ec_c",
        "ec_d",
    ]

    for k in keys_to_show:
        if k in meta_scalar.index:
            print(f"{k:25s}: {meta_scalar[k]}")

    print("\nAll scalar metadata columns:")
    print(meta_scalar)

print(f"\nAll figures saved in: {PLOT_DIR}")
```

```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from pathlib import Path


# ============================================================
# Settings: must match the Oxygen calculation script
# ============================================================

# Same value used in the calculation script
O2_TARGET_DECREASE_PERCENT = 1.5
target_tag = str(O2_TARGET_DECREASE_PERCENT).replace(".", "p")

AERAid = f"AERA_O2var_0_200m_minus{target_tag}pct_hist_nird"

FHIST_AERA_INPUT = Path(
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/"
    "TipESM/cmor/esm-up2p0/v20251010/AERA/input/references/"
    "NorESM2-LM_esm-hist_r1i1p1f1_1850-2014_AERA_O2var_0_200m_full_input.csv"
)

O2_COL = "O2var"

y0 = 1850
y2 = 2014
nyr = 5

ref_start = 1850
ref_end = 1899

AERA_DIR = Path(
    "/nird/datalake/NS2980K/users/yongyub/O2_linearlity/"
    "TipESM/cmor/esm-up2p0/v20251010/AERA"
)

OUTPUT_DIR = AERA_DIR / "output" / AERAid

FOUT = OUTPUT_DIR / f"{AERAid}_ann_O2var_co2_ems_{y2}.dat"
meta_file = OUTPUT_DIR / f"{AERAid}_meta_{y2}.nc"
AERA_outfile = OUTPUT_DIR / f"{AERAid}_{y2}.csv"
target_info_file = OUTPUT_DIR / f"{AERAid}_target_info_{y2}.csv"

PLOT_DIR = OUTPUT_DIR / f"figures_check_{y2}"
PLOT_DIR.mkdir(parents=True, exist_ok=True)


# ============================================================
# Helper functions
# ============================================================
def check_exists(path):
    if not Path(path).exists():
        raise FileNotFoundError(f"File does not exist: {path}")


def savefig(name):
    out = PLOT_DIR / name
    plt.tight_layout()
    plt.savefig(out, dpi=200, bbox_inches="tight")
    plt.show()
    print(f"Saved: {out}")


def first_existing_column(df, candidates):
    for c in candidates:
        if c in df.columns:
            return c
    return None


def first_existing_index(series, candidates):
    for c in candidates:
        if c in series.index:
            return c
    return None


def require_years(series, years, name):
    missing = years[~np.isin(years, series.index.values)]
    if len(missing) > 0:
        raise ValueError(
            f"{name} does not contain all required years. "
            f"Missing years include: {missing[:10]}"
        )


# ============================================================
# Check required files
# ============================================================
check_exists(FHIST_AERA_INPUT)
check_exists(FOUT)
check_exists(AERA_outfile)


# ============================================================
# 1. Read input CSV used for AERA-Oxygen calculation
# ============================================================
df_input = pd.read_csv(FHIST_AERA_INPUT, index_col="year")
df_input.index = df_input.index.astype(int)
df_input = df_input.sort_index()

input_years = np.arange(y0, y2 + 1)
ref_years = np.arange(ref_start, ref_end + 1)
next5_years = np.arange(y2 + 1, y2 + nyr + 1)

if O2_COL not in df_input.columns:
    raise ValueError(
        f"Input CSV does not contain '{O2_COL}'. "
        f"Available columns: {list(df_input.columns)}"
    )

if "ff_emission" not in df_input.columns:
    raise ValueError(
        f"Input CSV does not contain 'ff_emission'. "
        f"Available columns: {list(df_input.columns)}"
    )

input_o2 = pd.to_numeric(df_input.loc[input_years, O2_COL], errors="coerce")
input_ff = pd.to_numeric(df_input.loc[input_years, "ff_emission"], errors="coerce")

if input_o2.isna().any():
    bad = input_o2.index[input_o2.isna()]
    raise ValueError(f"{O2_COL} has NaN values for years: {bad[:10].tolist()}")

if input_ff.isna().any():
    bad = input_ff.index[input_ff.isna()]
    raise ValueError(f"ff_emission has NaN values for years: {bad[:10].tolist()}")


# ============================================================
# 2. Read / calculate oxygen target
# ============================================================
if target_info_file.exists():
    target_info = pd.read_csv(target_info_file)
    O2_TARGET_ABS = float(target_info.loc[0, "oxygen_target_abs"])
    O2_REF_MEAN = float(target_info.loc[0, "reference_mean"])
    target_decrease_percent_used = float(target_info.loc[0, "target_decrease_percent"])
else:
    O2_REF_MEAN = pd.to_numeric(
        df_input.loc[ref_years, O2_COL],
        errors="coerce",
    ).mean()
    O2_TARGET_ABS = O2_REF_MEAN * (1.0 - O2_TARGET_DECREASE_PERCENT / 100.0)
    target_decrease_percent_used = O2_TARGET_DECREASE_PERCENT

print("\nOxygen target information")
print("=" * 60)
print(f"Reference period             : {ref_start}-{ref_end}")
print(f"Reference mean {O2_COL}      : {O2_REF_MEAN:.10f}")
print(f"Target decrease              : {target_decrease_percent_used:.3f} %")
print(f"Absolute target {O2_COL}     : {O2_TARGET_ABS:.10f}")
print(f"Stocktake-year {O2_COL}      : {input_o2.loc[y2]:.10f}")


# ============================================================
# 3. Read pickle output
# ============================================================
with open(FOUT, "rb") as f:
    year, ann_o2var, ann_co2, ann_ems = pickle.load(f)

year = np.asarray(year)
ann_o2var = np.asarray(ann_o2var)
ann_ems = np.asarray(ann_ems)

valid_year = ~np.isnan(year)
year_valid = year[valid_year].astype(int)

o2_series = pd.Series(
    ann_o2var[valid_year],
    index=year_valid,
    name="ann_O2var",
).dropna()

ems_series = pd.Series(
    ann_ems[valid_year],
    index=year_valid,
    name="ann_ems",
).dropna()


# ============================================================
# 4. Read AERA-Oxygen future emission CSV
# ============================================================
aera_future = pd.read_csv(AERA_outfile, index_col=0)
aera_future.index = aera_future.index.astype(int)

if "ff_emission" in aera_future.columns:
    future_ff = aera_future["ff_emission"].copy()
else:
    future_ff = aera_future.iloc[:, 0].copy()

future_ff.name = "ff_emission"
future_ff = pd.to_numeric(future_ff, errors="coerce").dropna()

require_years(future_ff, next5_years, "AERA future emission CSV")
require_years(ems_series, next5_years, "pickle ann_ems")


# ============================================================
# 5. Read metadata CSVs if available
# ============================================================
timeseries_csv = Path(str(meta_file) + ".timeseries.csv")
scalar_csv = Path(str(meta_file) + ".scalar.csv")

meta_ts_run = None
meta_scalar = None

if timeseries_csv.exists():
    meta_ts = pd.read_csv(timeseries_csv, index_col=[0, 1])
    meta_ts.index.names = ["year_stocktake", "year"]

    meta_ts = meta_ts.apply(pd.to_numeric, errors="coerce")
    stocktake_years = meta_ts.index.get_level_values("year_stocktake").astype(int)

    if y2 in stocktake_years:
        meta_ts_run = meta_ts.loc[y2].copy()
        meta_ts_run.index = meta_ts_run.index.astype(int)

if scalar_csv.exists():
    meta_scalar_all = pd.read_csv(scalar_csv, index_col=0)
    meta_scalar_all.index = meta_scalar_all.index.astype(int)
    meta_scalar_all = meta_scalar_all.apply(pd.to_numeric, errors="coerce")

    if y2 in meta_scalar_all.index:
        meta_scalar = meta_scalar_all.loc[y2].copy()


# ============================================================
# Fig 1. Input O2var
# ============================================================
plt.figure(figsize=(9, 4.5))
plt.plot(input_o2.index, input_o2.values, lw=2, label="input O2var")
plt.axhline(
    O2_REF_MEAN,
    ls=":",
    lw=1.5,
    label=f"{ref_start}-{ref_end} mean",
)
plt.axhline(
    O2_TARGET_ABS,
    ls="--",
    lw=1.5,
    label=f"O2 target = {O2_TARGET_ABS:.4f}",
)
plt.axvline(y2, ls="--", lw=1, label=f"stocktake year = {y2}")
plt.xlabel("Year")
plt.ylabel("O$_2$ inventory 0–200 m (Pmol)")
plt.title("Input O2var used for AERA-Oxygen")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_01_input_O2var_{y2}.png")


# ============================================================
# Fig 2. Input ff_emission and next 5-year stored emissions
# ============================================================
plt.figure(figsize=(9, 4.5))
plt.plot(
    ems_series.index,
    ems_series.values,
    lw=2,
    marker="o",
    label="stored ff_emission in pickle",
)
plt.axvline(y2, ls="--", lw=1, label=f"stocktake year = {y2}")
plt.axvspan(y2 + 1, y2 + nyr, alpha=0.15, label="next 5 years")
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("Fossil-fuel CO$_2$ emission stored in pickle")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_02_pickle_ff_emission_{y2}.png")


# ============================================================
# Fig 3. Full AERA-Oxygen future emission curve
# ============================================================
plt.figure(figsize=(9, 4.5))

plt.plot(
    input_ff.index,
    input_ff.values,
    lw=2,
    label="input historical ff_emission",
)

plt.plot(
    future_ff.index,
    future_ff.values,
    lw=2,
    label="AERA-Oxygen future ff_emission curve",
)

plt.scatter(
    next5_years,
    future_ff.loc[next5_years],
    s=50,
    zorder=5,
    label="next 5 years used",
)

plt.axvline(y2, ls="--", lw=1)
plt.axhline(0, lw=1)
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("AERA-Oxygen-calculated fossil-fuel emission curve")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_03_aera_oxygen_future_ff_curve_{y2}.png")


# ============================================================
# Fig 4. Zoom around stocktake
# ============================================================
zoom_start = max(y0, y2 - 25)
zoom_end = y2 + 40

plt.figure(figsize=(9, 4.5))

plt.plot(
    input_ff.loc[zoom_start:y2].index,
    input_ff.loc[zoom_start:y2].values,
    lw=2,
    marker="o",
    label="input historical ff_emission",
)

plt.plot(
    future_ff.loc[y2 + 1:zoom_end].index,
    future_ff.loc[y2 + 1:zoom_end].values,
    lw=2,
    marker="o",
    label="AERA-Oxygen future curve",
)

plt.scatter(
    next5_years,
    future_ff.loc[next5_years],
    s=60,
    zorder=5,
    label="next 5 years used",
)

plt.axvline(y2, ls="--", lw=1)
plt.axhline(0, lw=1)
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("AERA-Oxygen emission curve near stocktake")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_04_zoom_ff_curve_{y2}.png")


# ============================================================
# Fig 5. Stored next-5-year emissions vs AERA_outfile
# ============================================================
pickle_next5 = ems_series.loc[next5_years]
csv_next5 = future_ff.loc[next5_years]
diff_next5 = pickle_next5 - csv_next5

plt.figure(figsize=(9, 4.5))
plt.plot(next5_years, pickle_next5, marker="o", lw=2, label="pickle ann_ems")
plt.plot(next5_years, csv_next5, marker="x", lw=2, label="AERA-Oxygen csv")
plt.xlabel("Year")
plt.ylabel("ff emission (PgC yr$^{-1}$)")
plt.title("Check: next 5-year emissions in pickle vs AERA-Oxygen CSV")
plt.grid(alpha=0.3)
plt.legend()
savefig(f"{AERAid}_check_05_next5_pickle_vs_csv_{y2}.png")

print("\nNext 5-year emission check")
print(pd.DataFrame({
    "pickle_ann_ems": pickle_next5,
    "aera_oxygen_csv": csv_next5,
    "difference": diff_next5,
}))


# ============================================================
# Fig 6. Cumulative fossil-fuel emissions
# ============================================================
combined_ff = pd.concat(
    [
        input_ff.loc[:y2],
        future_ff.loc[future_ff.index > y2],
    ]
).sort_index()

plt.figure(figsize=(9, 4.5))
plt.plot(combined_ff.index, combined_ff.cumsum(), lw=2)
plt.axvline(y2, ls="--", lw=1)
plt.xlabel("Year")
plt.ylabel("Cumulative ff emission (PgC)")
plt.title("Cumulative fossil-fuel CO$_2$ emissions")
plt.grid(alpha=0.3)
savefig(f"{AERAid}_check_06_cumulative_ff_emission_{y2}.png")


# ============================================================
# Fig 7. Metadata total_emission and ff_emission
# ============================================================
if meta_ts_run is not None:
    cols = [c for c in ["total_emission", "ff_emission"] if c in meta_ts_run.columns]

    if len(cols) > 0:
        plt.figure(figsize=(9, 4.5))

        for col in cols:
            plt.plot(
                meta_ts_run.index,
                meta_ts_run[col],
                lw=2,
                label=col,
            )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("Emission (PgC yr$^{-1}$)")
        plt.title("AERA-Oxygen metadata: total emission and ff emission")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_07_meta_total_vs_ff_emission_{y2}.png")


# ============================================================
# Fig 8. Metadata emission decomposition
# total_emission - ff_emission = lu_emission + non_co2_emission
# ============================================================
if meta_ts_run is not None:
    if ("total_emission" in meta_ts_run.columns) and ("ff_emission" in meta_ts_run.columns):
        residual = meta_ts_run["total_emission"] - meta_ts_run["ff_emission"]

        plt.figure(figsize=(9, 4.5))
        plt.plot(
            meta_ts_run.index,
            meta_ts_run["ff_emission"],
            lw=2,
            label="ff_emission",
        )
        plt.plot(
            meta_ts_run.index,
            residual,
            lw=2,
            label="lu + nonCO2 contribution",
        )
        plt.plot(
            meta_ts_run.index,
            meta_ts_run["total_emission"],
            lw=2,
            label="total_emission",
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("Emission (PgC yr$^{-1}$)")
        plt.title("AERA-Oxygen metadata: emission decomposition")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_08_meta_emission_decomposition_{y2}.png")


# ============================================================
# Fig 9. Oxygen diagnostic from metadata
# ============================================================
if meta_ts_run is not None:
    oxygen_anth_col = first_existing_column(
        meta_ts_run,
        [
            "oxygen_anth",
            "O2var_anth",
            "o2_anth",
            "O2_anth",
            "temp_anth",  # fallback
        ],
    )

    if oxygen_anth_col is not None:
        plt.figure(figsize=(9, 4.5))

        plt.plot(
            input_o2.index,
            input_o2.values,
            lw=1.5,
            alpha=0.6,
            label="input O2var",
        )

        plt.plot(
            meta_ts_run.index,
            meta_ts_run[oxygen_anth_col],
            lw=2.5,
            label=f"AERA-Oxygen {oxygen_anth_col}",
        )

        target_key = None
        if meta_scalar is not None:
            target_key = first_existing_index(
                meta_scalar,
                [
                    "oxygen_target_abs",
                    "O2var_target_abs",
                    "o2_target_abs",
                    "temp_target_abs",  # fallback
                ],
            )

        if target_key is not None:
            plt.axhline(
                meta_scalar[target_key],
                ls="--",
                lw=1.5,
                label=f"target from metadata ({target_key})",
            )
        else:
            plt.axhline(
                O2_TARGET_ABS,
                ls="--",
                lw=1.5,
                label=f"O2 target = {O2_TARGET_ABS:.4f}",
            )

        plt.axhline(
            O2_REF_MEAN,
            ls=":",
            lw=1.2,
            label=f"{ref_start}-{ref_end} mean",
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.xlabel("Year")
        plt.ylabel("O$_2$ inventory 0–200 m (Pmol)")
        plt.title("AERA-Oxygen O2var diagnostic")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_09_O2var_diagnostic_{y2}.png")


# ============================================================
# Fig 10. Relative / target-gap Oxygen diagnostic if available
# ============================================================
if meta_ts_run is not None:
    oxygen_rel_col = first_existing_column(
        meta_ts_run,
        [
            "oxygen_anth_rel",
            "O2var_anth_rel",
            "o2_anth_rel",
            "O2_anth_rel",
            "temp_anth_rel",  # fallback
        ],
    )

    if oxygen_rel_col is not None:
        plt.figure(figsize=(9, 4.5))

        plt.plot(
            meta_ts_run.index,
            meta_ts_run[oxygen_rel_col],
            lw=2,
            label=oxygen_rel_col,
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)

        plt.xlabel("Year")
        plt.ylabel("Relative / target-gap O2var")
        plt.title("AERA-Oxygen relative O2var diagnostic")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_10_O2var_relative_diagnostic_{y2}.png")


# ============================================================
# Fig 11. Reconstruct cubic curve from metadata coefficients
# ============================================================
if meta_scalar is not None:
    required = ["ec_a", "ec_b", "ec_c", "ec_d", "ec_target_year_rel"]

    if all(k in meta_scalar.index for k in required):
        a = meta_scalar["ec_a"]
        b = meta_scalar["ec_b"]
        c = meta_scalar["ec_c"]
        d = meta_scalar["ec_d"]
        target_year_rel = int(meta_scalar["ec_target_year_rel"])

        t = np.arange(0, target_year_rel + 1)
        recon_year = y2 + t
        recon_ff = a * t**3 + b * t**2 + c * t + d

        plt.figure(figsize=(9, 4.5))

        plt.plot(
            recon_year,
            recon_ff,
            lw=2.5,
            label="reconstructed cubic curve",
        )

        plt.plot(
            future_ff.index,
            future_ff.values,
            lw=1.5,
            ls="--",
            label="saved AERA-Oxygen future ff_emission",
        )

        plt.scatter(
            next5_years,
            future_ff.loc[next5_years],
            s=60,
            zorder=5,
            label="next 5 years used",
        )

        plt.axvline(y2, ls="--", lw=1)
        plt.axhline(0, lw=1)
        plt.xlabel("Year")
        plt.ylabel("ff emission (PgC yr$^{-1}$)")
        plt.title("Reconstructed AERA-Oxygen cubic emission curve")
        plt.grid(alpha=0.3)
        plt.legend()
        savefig(f"{AERAid}_check_11_reconstructed_cubic_curve_{y2}.png")


# ============================================================
# Print scalar diagnostics
# ============================================================
if meta_scalar is not None:
    print("\nAERA-Oxygen scalar diagnostics")
    print("=" * 60)

    keys_to_show = [
        "oxygen_target_abs",
        "O2var_target_abs",
        "o2_target_abs",
        "temp_target_abs",
        "total_emission_budget",
        "ff_emission_budget",
        "reb",
        "ec_cost",
        "ec_reb_diff",
        "ec_overshoot_integral",
        "ec_overshoot",
        "ec_curvature",
        "ec_target_year_rel",
        "ec_a",
        "ec_b",
        "ec_c",
        "ec_d",
    ]

    for k in keys_to_show:
        if k in meta_scalar.index:
            print(f"{k:25s}: {meta_scalar[k]}")

    print("\nAll scalar metadata columns:")
    print(meta_scalar)

print(f"\nAll figures saved in: {PLOT_DIR}")
```
