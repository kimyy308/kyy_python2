```python
#--Importing Modules
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cmocean
from cmcrameri import cm
```

```python
#--Reading Dataset
ds1=xr.open_dataset('/proj/jedwards/archive/b.e21.BHISTcmip6.f09_g17.LE2-1001.001/atm/proc/tseries/month_1/b.e21.BHISTcmip6.f09_g17.LE2-1001.001.cam.h0.PRECT.195001-195912.nc') 
ds2=xr.open_dataset('/proj/jedwards/archive/b.e21.BHISTcmip6.f09_g17.LE2-1001.001/atm/proc/tseries/month_1/b.e21.BHISTcmip6.f09_g17.LE2-1001.001.cam.h0.U.195001-195912.nc') 
ds3=xr.open_dataset('/proj/jedwards/archive/b.e21.BHISTcmip6.f09_g17.LE2-1001.001/atm/proc/tseries/month_1/b.e21.BHISTcmip6.f09_g17.LE2-1001.001.cam.h0.V.195001-195912.nc') 
```

```python
#--Extracting varibales as per requirement
prec = (ds1.PRECT*8.64e+7).squeeze().load()  # Precipitation 
uvel = (ds2.U).isel(lev=25).squeeze().load() # 850 hpa Zonal Wind 
vvel = (ds3.V).isel(lev=25).squeeze().load() # 850 hpa Meridional Wind
```

```python
#---Annual Mean Climatology
prec_mean = prec.groupby('time.year').mean().mean('year')  # Precipitation mean
uvel_mean = uvel.groupby('time.year').mean().mean('year')  # 850 hpa Zonal Wind Mean
vvel_mean = vvel.groupby('time.year').mean().mean('year')  # 850 hpa Meridional Wind Mean

```

```python
#---JJAS Climatology
prec_jjas = prec.where(prec.time.dt.month.isin([6,7,8,9]), drop=True).groupby('time.year').mean().mean('year')  # Precipitation JJAS
uvel_jjas = uvel.where(uvel.time.dt.month.isin([6,7,8,9]), drop=True).groupby('time.year').mean().mean('year') # 850 hpa Zonal JJAS
vvel_jjas = vvel.where(vvel.time.dt.month.isin([6,7,8,9]), drop=True).groupby('time.year').mean().mean('year') # 850 hpa Meridional Wind JJAS
```

```python
#-- Plotting Mean Precipitation
plt.rcParams["font.size"] = 12
proj=ccrs.Robinson(central_longitude=180)
fig=plt.figure(figsize=(18,7))
#----------------------------
ax=plt.subplot(121, projection=proj)
ax.add_feature(cfeature.LAND,linewidth=0.5, color='lightgrey')
ax.coastlines()
#ax.set_extent((-180, 180, -90, 90), crs=ccrs.PlateCarree())
plt.contourf(prec_mean.lon, prec_mean.lat, prec_mean, np.arange(0,20,0.5), cmap=cmocean.cm.rain, extend='max', transform=ccrs.PlateCarree())
cbar=plt.colorbar(orientation="horizontal", pad=0.01, shrink=0.6,aspect=35)
plt.title('Annual Mean Precipitation')
cbar.set_label(label='mm/day')
#----------------------------
ax=plt.subplot(122, projection=proj)
ax.add_feature(cfeature.LAND,linewidth=0.5, color='lightgrey')
ax.coastlines()
#ax.set_extent((-180, 180, -90, 90), crs=ccrs.PlateCarree())
plt.contourf(prec_jjas.lon, prec_jjas.lat, prec_jjas, np.arange(0,20,0.5), cmap=cmocean.cm.rain, extend='max', transform=ccrs.PlateCarree())
cbar=plt.colorbar(orientation="horizontal", pad=0.01, shrink=0.6,aspect=35)
plt.title('Mean JJAS Precipitation')
cbar.set_label(label='mm/day')

plt.tight_layout()
```

```python
#-- Marnitude of Wind Speed
mag_wind = np.sqrt(uvel_jjas**2 + vvel_jjas**2)
```

```python
#-- 850 hpa Wind overlay on Mean Precipitation
plt.rcParams["font.size"] = 12
proj=ccrs.Robinson(central_longitude=180)
fig=plt.figure(figsize=(18,7), dpi=300)
#----------------------------
ax=plt.subplot(121, projection=proj)
ax.add_feature(cfeature.LAND,linewidth=0.5, color='lightgrey')
ax.coastlines()
#ax.set_extent((-180, 180, -90, 90), crs=ccrs.PlateCarree())
plt.contourf(mag_wind.lon, mag_wind.lat, mag_wind, np.arange(0,20,0.5), cmap=cmocean.cm.matter, extend='max', transform=ccrs.PlateCarree())
cbar=plt.colorbar(orientation="horizontal", pad=0.01, shrink=0.6,aspect=35)
cbar.set_label(label='m/s')

Q=plt.quiver(uvel_jjas.lon.values[::4], uvel_jjas.lat.values[::4], uvel_jjas.values[::4, ::4], vvel_jjas.values[::4, ::4], scale=250, color='black', width=0.0015, transform=ccrs.PlateCarree())
qk=plt.quiverkey(Q,0.9,-0.1,8,'10 m/s',labelpos='N')
```

```python
## Indian Monsoon and global Precipitation
prec_im = prec.sel(lat=slice(8,30), lon=slice(50,90)).mean(('lat', 'lon')).where(prec.time.dt.month.isin([6,7,8,9]), drop=True).groupby('time.year').mean()
prec_gl = prec.mean(('lat', 'lon')).groupby('time.year').mean()
```

```python
#-- Simple Time Series plot
fig=plt.figure(figsize=(18,7), dpi=300)
plt.plot(prec_im.year, prec_im, color='red', lw=3, label='India Rainfall')
plt.plot(prec_gl.year, prec_gl, color='black', lw=3, label='Global rainfall')

plt.legend(fancybox=False, framealpha=0, shadow=False, borderpad=0.5, ncol=1, loc="upper right", prop={'size': 20})
plt.ylabel('mm/day')
plt.xlabel('years')
```
