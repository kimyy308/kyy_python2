```python
import cdsapi
import os

# ORAS5 Gridded 데이터셋 원본 선언
dataset = "reanalysis-oras5"
client = cdsapi.Client()

years = [str(y) for y in range(1982, 2021)]
months = [f"{m:02d}" for m in range(1, 13)]

print("Starting annual data download loop (Global Grid to be subsetted locally)...")

for year in years:
    # 전구 데이터를 받은 뒤 로컬에서 자를 예정이므로 파일명 변경
    output_filename = f"oras5_sst_global_{year}.nc"
    
    if os.path.exists(output_filename):
        print(f"[{year}] File {output_filename} already exists. Skipping...")
        continue
        
    print(f"[{year}] Requesting global grid data for year {year}...")
    
    # -------------------------------------------------------------------------
    # [완벽 교정] 서버가 지원하지 않는 "area" 파라미터를 완전히 제거했습니다.
    # 단일 레벨 SST는 전구 데이터를 다 받아도 파일당 15MB 내외로 매우 가볍습니다.
    # -------------------------------------------------------------------------
    request = {
        "product_type": ["consolidated", "operational"],
        "vertical_resolution": "single_level",
        "variable": ["sea_surface_temperature"],
        "year": [year],
        "month": months
    }
    
    try:
        client.retrieve(dataset, request).download(output_filename)
        print(f"[{year}] Successfully downloaded global {output_filename}")
    except Exception as e:
        print(f"🚨 [{year}] Failed to download due to error: {e}")

print("All global files downloaded successfully.")
```
